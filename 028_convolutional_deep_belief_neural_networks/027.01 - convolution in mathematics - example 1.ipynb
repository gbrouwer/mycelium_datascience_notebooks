# Convolution on discrete timeseries
Convolution is a mathematical operation on two function (f and g) that produces a third function that expresses how the shape of one is modified by the other. More specifically, the resulting function is created by progressively shifting a reversed (-f) function f relative to function g (or vice versa), computing the product of the two and taking the integral, creating the resulting function h. This is mathematically correct, but we imagine hopelessely abstract for many of our readers. Therefore, let us take two discrete timeseries f and g.
In most applications of convolution, g will be much shorter than f. In these cases f does still represent a series of measurements. However, the shorter g usually takes on a charactistic shape not based on any measurement, but instead reflecting some operation a system applies to a timeseries passing through it. In this case we refer to g as a kernel, an impulse response function, or a filter. 

Convolution has a variety of practical application. For example, many effects applied to many forms of modern music can be seen as applying certain filters to the audio. Delay, echo, reverb, low/high/bandpass filtering, even basic settings like bass, treble and high can in fact be seen in some ways as convolutions of a filter g on a timeseries f. Similarly, in photography, things like blurring, sharpening and edge detection can also be imagined as applying a small 2D filter or kernel g to a much large image f. In probability theory, the probability distribution of the sum of two independent random variables is the convolution of their individual distributions. As a final example, the polarization of a neuron in responsed to a brief impulse of inputs will decay back to baseline in way that can be computed by convolving the impulse (as a timeseries) and the impulse response function of a typical neuron. If the inputs combine into a strong enough polarization, the neuron sets of a chain of reactions, all governed by their own impulse response functions that has the neuron produce an action potential: an impulse send to neurons connecting to it.

In many instances, convolution is very similar to computing a cross correlation, with the only difference being that in convolution, the filter g is reversed. In other words, the cross correlation of f(x) and f(g) is the convolution of f(x) and g(-x). Practically, this means that before we compute the convolution between f and g we need to reverse g, after we have padded it to make it the same length as f. Although it is not immedtionaly obvious in this notebook why this reversal is needed, it will be become clear in the next notebook in which we model our filter as a asymmetric responde function. 

The above is likely to still be abstract, so let us examine an example. We will explicitely code the convolution function in pure python, before switching to inbuilt function that come with SciPy and Numpy. In the example below, time series f is all 0s at the beginning and end, with a brief sequence of 1s in the middle. You can think of this as an 'on signal' or a state change of the system from one value into another, or perhaps a pulse of inputs lasting as long as we see a 1 in the timeseries, before returning ito the off state of zeros. The second timeseries is a normal/gaussian distribution, normalized to a maximum of 1, padded by a significant sequence of 0s left and right of its mean centered midway the timeseries g. After making g equal in length to f, we add additional padding to both f and g, to both sides, equal to the length of filter g. In this way we can incrementally shift the filter g from its starting point completely outside of timeseries f in the area padded with zeros 0 all the way to it being completely outside series f in the area padded with zeros 0 on the other side (the two gray regions in the image). 

### Generate and Plot the two timeseries f and g

```python
#Block Function f
f = np.zeros((1401,1))
f[600:801,:] = 1

#Normal Distribution g
xval = np.linspace(-200,1200,1401)
gdistr = mnorm(500,500)
g = gdistr.pdf(xval)
g = g / np.max(g)
g = np.expand_dims(g,-1)[::-1]
xticks = np.linspace(-200,1200,15).astype(int).astype('str')
xticks[0] = ''
xticks[1] = ''
xticks[-2] = ''
xticks[-1] = ''
padding0 = np.array([[-200,0], [0,0], [0,1], [-200,1], [-200,0]])
padding1 = np.array([[1000,0], [1200,0], [1200,1], [1000,1], [1000,0]])
p0 = Polygon(padding0, facecolor = (0.5,0.5,0.5),alpha=0.25)
p1 = Polygon(padding1, facecolor = (0.5,0.5,0.5),alpha=0.25)

#Plot f and g
fig = plt.figure(figsize=(12,8));
plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.5)
ax = plt.subplot(2,1,1);
plt.plot(xval,f,color=colors[1,:],linewidth=3);
plt.xticks(np.linspace(-200,1200,15),xticks,**csfont,fontsize=18);
plt.yticks(np.linspace(0,1,5),np.round(np.linspace(0,1,5),2),**csfont,fontsize=18);
plt.grid('on');
plt.title('f(x)',**csfont,fontsize=32,pad=20);
ax.add_patch(p0);
ax.add_patch(p1);

padding0 = np.array([[-200,0], [0,0], [0,1], [-200,1], [-200,0]])
padding1 = np.array([[1000,0], [1200,0], [1200,1], [1000,1], [1000,0]])
p0 = Polygon(padding0, facecolor = (0.5,0.5,0.5),alpha=0.25)
p1 = Polygon(padding1, facecolor = (0.5,0.5,0.5),alpha=0.25)

#Plot f and g
plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.5)
ax = plt.subplot(2,1,2);
plt.plot(xval,g,color=colors[3,:],linewidth=3);
plt.xticks(np.linspace(-200,1200,15),xticks,**csfont,fontsize=18);
plt.yticks(np.linspace(0,1,5),np.round(np.linspace(0,1,5),2),**csfont,fontsize=18);
plt.grid('on')
plt.title('g(x)',**csfont,fontsize=32,pad=20);
ax.add_patch(p0);
ax.add_patch(p1);
plt.savefig('../../assets/matplotlib/26-conv1.png', dpi=300,bbox_inches="tight")
```

### Pure Python Convolution - Init
The loop below implements the cross correlation between two time series. This is achieves by rolling one function (typically the second shorter function representing the impulse reponse function or kernel) incrementally. First, we shift the function currently centered in the exact middle all the way rightward so that the entire non-zero part of g is outside of f, but instead in the right-side padded region. 

```python
g = np.roll(g,-600)
padding0 = np.array([[-200,0], [0,0], [0,1], [-200,1], [-200,0]])
padding1 = np.array([[1000,0], [1200,0], [1200,1], [1000,1], [1000,0]])
p0 = Polygon(padding0, facecolor = (0.5,0.5,0.5),alpha=0.25)
p1 = Polygon(padding1, facecolor = (0.5,0.5,0.5),alpha=0.25)

plt.figure(figsize=(12,8));
plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.5)
ax = plt.subplot(2,1,1);
plt.plot(xval,f,color=colors[1,:],linewidth=3);
plt.xticks(np.linspace(-200,1200,15),xticks,**csfont,fontsize=18);
plt.yticks(np.linspace(0,1,5),np.round(np.linspace(0,1,5),2),**csfont,fontsize=18);
plt.grid('on')
ax.add_patch(p0)
ax.add_patch(p1)
plt.title('f(x)',**csfont,fontsize=32,pad=20);

padding0 = np.array([[-200,0], [0,0], [0,1], [-200,1], [-200,0]])
padding1 = np.array([[1000,0], [1200,0], [1200,1], [1000,1], [1000,0]])
p0 = Polygon(padding0, facecolor = (0.5,0.5,0.5),alpha=0.25)
p1 = Polygon(padding1, facecolor = (0.5,0.5,0.5),alpha=0.25)

ax = plt.subplot(2,1,2);
plt.plot(xval,g,color=colors[3,:],linewidth=3);
plt.xticks(np.linspace(-200,1200,15),xticks,**csfont,fontsize=18);
plt.yticks(np.linspace(0,1,5),np.round(np.linspace(0,1,5),2),**csfont,fontsize=18);
plt.grid('on')
ax.add_patch(p0)
ax.add_patch(p1)
plt.title('g(x)',**csfont,fontsize=32,pad=20);
plt.savefig('../../assets/matplotlib/26-conv2.png', dpi=300,bbox_inches="tight")
```

### Pure Python Convolution - Compute Integral
After initialization, we compute h as the product of f and g' (the shifted version of g). By summing all values of the resulting timeseries we obtain the area of the overlap between f and g', or integral. This value is stored in h at index 0. We then continue to shift g one index to the right and repeat the same computation until we have shifted the curve back to its started point. 

```python
h = []
rval = np.linspace(-200,1200,1401)
for i in range(rval.shape[0]):
    g_prime = np.roll(g,-int(rval[i]))
    integral = np.sum(g_prime[:,0]*f[:,0])
    h.append(integral)
h = np.array(h)
h = h / np.max(h)
h = h.copy()
h = np.roll(h,-200)
```
