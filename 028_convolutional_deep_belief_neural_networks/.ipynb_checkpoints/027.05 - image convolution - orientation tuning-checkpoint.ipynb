### Gabor Patch Filter

```python
x = np.linspace(-10,10,1001)
X, Y = np.meshgrid(x, x)
XX = X.flatten()
YY = Y.flatten()
XY = np.vstack((XX,YY)).T
gauss_distr = mnorm([0,0],[[15,0],[0,15]])
G = gauss_distr.pdf(XY)
G = np.reshape(G,((1001,1001)))
G[G < 0.0001] = 0
Gdict = {}

wavelengths = np.linspace(3,13,4)
orientations = np.linspace(0,np.pi,19)
for w,wavelength in enumerate(wavelengths):
    for o,orientation in enumerate(orientations):
        angle = orientation
        grating = np.sin(2*np.pi*(X*np.sin(angle) + Y*np.cos(angle)) / wavelength)
        grating = grating*G
        gaborim = grating - np.min(grating)
        gaborim = gaborim / np.max(gaborim) * 255
        gaborim = gaborim.astype('int8');
        gaborim = Image.fromarray(gaborim);
        gaborim = gaborim.resize((32,32)) 
        grating = np.array(gaborim) / 256;
        Gdict[(w,o)] = grating - 0.5
```

### Draw all orientation and wavelengths

```python
count = 0
plt.figure(figsize=(14,14))
for o,orientation in enumerate(orientations):
    for w,wavelength in enumerate(wavelengths):
        count = count + 1
        gabor = Gdict[(w,o)]
        plt.subplot(19,4,count)
        plt.imshow(gabor,cmap='gray')
        plt.axis('off')
```

### Compute Response

```python
count = 0
tuning = Gdict[(0,10)]
# plt.figure(figsize=(14,14))
R = np.zeros((19,4))
for o,orientation in enumerate(orientations):
    for w,wavelength in enumerate(wavelengths):
        count = count + 1
        gabor = Gdict[(w,o)]
        response = np.sum(convolve2d(gabor,tuning,'same'));
        R[o,w] = response
R = R - np.min(R)
R = R / np.max(R)
```
