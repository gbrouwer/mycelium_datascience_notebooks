### Import

```python
import os
import sys
import numpy as np
import pandas as pd
import imageio
import pickle
import json
import matplotlib.pyplot as plt
import pickle
from tqdm import tqdm
from PIL import Image
from scipy.spatial.distance import pdist
from scipy.spatial.distance import cdist
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Find All Images

```python
imagefiles = []
for root, dirs, files in os.walk("../../data/natural_images/", topdown=False):
    for name in files:
        if '.jpg' in name:
            imagefiles.append(os.path.join(root, name))
```

### Loop through images, each time generating 

```python
data = []
for i in range(len(imagefiles)):
    im = np.array(Image.open(imagefiles[i]))
    for j in range(800):
        x1 = np.random.randint(im.shape[0])
        y1 = np.random.randint(im.shape[1])    
        x2 = np.random.randint(im.shape[0])
        y2 = np.random.randint(im.shape[1])  
        pixel1 = (np.expand_dims(im[x1,y1,:],-1) / 255.0).T
        pixel2 = (np.expand_dims(im[x2,y2,:],-1) / 255.0).T
        distance = np.sqrt((x1-x2)**2 + (y1-y2)**2)
        euclidean = cdist(pixel1,pixel2,metric='euclidean')[0][0]
        cosine = cdist(pixel1,pixel2,metric='cosine')[0][0]
        data.append((i,distance,euclidean,cosine,pixel1[0,0],pixel1[0,1],pixel1[0,2],pixel2[0,0],pixel2[0,1],pixel2[0,2]))
data = np.array(data)
```

### Save

```python
X = {}
X['data'] = data
pickle.dump( X, open( "../../data/natural_images/distances.p", "wb" ) )
```

### Load Back In

```python
X = pickle.load(open("../../data/natural_images/distances.p", "rb"))
data = X['data']
```

### Round and remove anythin with a distance above 300

```python
data[:,1] = np.round(data[:,1])
ind = np.where(data[:,1] <= 120)[0]
data = data[ind,:]
```

### Create Histogram

```python
euclidean_hist = np.zeros((120,1))
cosine_hist = np.zeros((120,1))
for i in range(120):
    ind = np.where(data[:,1] == i)[0]
    euclidean_hist[i,0] = np.mean(data[ind,2])
    cosine_hist[i,0] = np.nanmean(data[ind,3])
```

### Plot

```python
plt.figure(figsize=(12,6))
plt.subplots_adjust(left=0.1,bottom=0.1, right=1.0, top=0.9, wspace=0.5, hspace=0.5)
plt.subplot(1,2,1)
plt.plot(euclidean_hist[:,0],'k',linewidth=2)
plt.grid();
plt.xlabel('Interpixel Distance',**csfont,fontsize=20,labelpad=20);
plt.ylabel('Euclidean distance',**csfont,fontsize=20);
plt.yticks(np.linspace(0,0.6,7),np.round(np.linspace(0,0.6,7),2).astype(float),**csfont,fontsize=16);
plt.xticks(np.linspace(0,120,7),np.linspace(0,120,7).astype(int),**csfont,fontsize=16);
plt.axis((0,120,0,0.6));

plt.subplot(1,2,2);
plt.plot(cosine_hist[:,0],'k',linewidth=2)
plt.grid();
plt.xlabel('Interpixel Distance',**csfont,fontsize=20);
plt.ylabel('Cosine dissimilarity',**csfont,fontsize=20,labelpad=20);
plt.yticks(np.linspace(0,0.05,5),np.round(np.linspace(0,0.05,5),2).astype(float),**csfont,fontsize=16);
plt.xticks(np.linspace(0,120,7),np.linspace(0,120,7).astype(int),**csfont,fontsize=16);
plt.axis((0,120,0,0.05));
plt.savefig('../../assets/matplotlib//026-natural-images.png', dpi=300,bbox_inches="tight");
```
