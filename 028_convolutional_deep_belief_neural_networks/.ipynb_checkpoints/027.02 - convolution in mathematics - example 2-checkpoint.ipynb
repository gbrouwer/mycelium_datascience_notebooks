# The Hemodynamic Response Function (HRF)
The previous example used a symmetrical kernel or impulse response function for g - a normal distribution of mean mu and standard deviation sigma. Because of this symmetry, we didn't had to worry about directionality. When we use an assymmetric function, however, we need to tweak our filter a bit before the actual convolution is computed. We will get to that in a bit. But first, let us define such an asymmetrical filter, one taking from neuroimaging research: the hemodynamic response function. In functional Magnetic Resonance Imaging, local neural activity is measured as the ratio between oxygenated and de-oxygenated blood. The underlying mechanism is not completely understood but the consensus is that neural activity is a metabolic process like all other bodily functions, and therefore requires oxygen as energy. Local neural activity increased the demand for oxygen, and the ratio between oxygenated and de-oxygenated blood shifts. However, this is not instantenious. When a brain area is activated, for example the visual cortex by showing subjects in the scanner images, instead of absolute darkness, it will take some time for the ratio in oxygen to shift. This sluggish response is captured by the HRF. For a brief presentation of an image (say 1 second), the rise and fall of the ratio can take up to 25 seconds to return to its previous baseline. In addition, there is a characteristic undershoot, where for some time the ratio actually reverse, before returning to baseline. 

```python
times = np.linspace(0,24,401)
peak_values = gamma.pdf(times, 6)
undershoot_values = gamma.pdf(times, 12)
hrf = peak_values - 0.35 * undershoot_values
hrf = hrf / np.max(hrf)
plt.figure(figsize=(8,6))
plt.rc('axes', axisbelow=True)
plt.plot(times,hrf,'k',linewidth=3)
plt.xticks(np.linspace(0,24,9),**csfont,fontsize=16)
plt.yticks(np.linspace(-0.2,1.0,7),**csfont,fontsize=16)
plt.grid()
plt.ylabel('activity (arbitrary units)',**csfont,fontsize=24)
plt.xlabel('time (seconds)',**csfont,fontsize=24)
plt.savefig('../../assets/matplotlib/26-hrf.png', dpi=300,bbox_inches="tight")
```

### Generate and Plot the two timeseries f and g

```python
#Pulse Function f
f = np.zeros((1800,1))
f[600:701,:] = 1
f[1000:1101,:] = 0.5

#HRF g
xval = np.linspace(-400,1400,1800)
g = np.zeros((1800,1))
g[0:401,0] = hrf[::-1]
xticks = np.linspace(-400,1400,10).astype(int).astype('str')
xticks[0] = ''
xticks[1] = ''
xticks[-2] = ''
xticks[-1] = ''
padding0 = np.array([[-400,0], [0,0], [0,1], [-400,1], [-400,0]])
padding1 = np.array([[1000,0], [1400,0], [1400,1], [1000,1], [1000,0]])
p0 = Polygon(padding0, facecolor = (0.5,0.5,0.5),alpha=0.25)
p1 = Polygon(padding1, facecolor = (0.5,0.5,0.5),alpha=0.25)

#Plot f and g
fig = plt.figure(figsize=(12,8));
plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.5)
ax = plt.subplot(2,1,1);
plt.plot(xval,f,color=colors[1,:],linewidth=3);
plt.xticks(np.linspace(-400,1400,10),xticks,**csfont,fontsize=18);
plt.yticks(np.linspace(0,1,5),np.round(np.linspace(0,1,5),2),**csfont,fontsize=18);
plt.grid('on');
plt.title('f(x)',**csfont,fontsize=32,pad=20);
ax.add_patch(p0);
ax.add_patch(p1);

padding0 = np.array([[-400,0], [0,0], [0,1], [-400,1], [-400,0]])
padding1 = np.array([[1000,0], [1400,0], [1400,1], [1000,1], [1000,0]])
p0 = Polygon(padding0, facecolor = (0.5,0.5,0.5),alpha=0.25)
p1 = Polygon(padding1, facecolor = (0.5,0.5,0.5),alpha=0.25)

#Plot f and g
plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.5)
ax = plt.subplot(2,1,2);
plt.plot(xval,g,color=colors[3,:],linewidth=3);
plt.xticks(np.linspace(-400,1400,10),xticks,**csfont,fontsize=18);
plt.yticks(np.linspace(0,1,5),np.round(np.linspace(0,1,5),2),**csfont,fontsize=18);
plt.grid('on');
plt.title('g(x)',**csfont,fontsize=32,pad=20);
ax.add_patch(p0);
ax.add_patch(p1);
plt.savefig('../../assets/matplotlib/26-conv3.png', dpi=300,bbox_inches="tight")
```

### Pure Python Convolution - Compute Integral
After initialization, we compute h as the product of f and g' (the shifted version of g). By summing all values of the resulting timeseries we obtain the area of the overlap between f and g', or integral. This value is stored in h at index 0. We then continue to shift g one index to right and repeat the same computation until we have shifted the curve back to its started point. 

### Compute Integral

```python
h = []
rval = np.linspace(-400,1400,1801)
for i in range(rval.shape[0]):
    g_prime = np.roll(g,-int(rval[i]))
    integral = np.sum(g_prime[:,0]*f[:,0])
    h.append(integral)
h = np.array(h)
h = h / np.max(h)
h = h.copy()
h = np.roll(h,-800)[::-1]
```
