### Load

```python
im_lenna = Image.open('../../data/test_images/lenna.png')
im_lenna = np.array(im_lenna) / 255.0
im_lenna_bw = np.mean(im_lenna,axis=2)

im_amsterdam = Image.open('../../data/test_images/architecture_2.png')
im_amsterdam = im_amsterdam.resize((512,512))
im_amsterdam = np.array(im_amsterdam) / 255.0
im_amsterdam_bw = np.mean(im_amsterdam,axis=2)

im_spiral = Image.open('../../data/test_images/spiral.png')

im_spiral = np.array(im_spiral) / 255.0
im_spiral_bw = np.mean(im_spiral,axis=2)

plt.figure(figsize=(14,14));
plt.subplot(2,2,1)
plt.imshow(im_lenna_bw,cmap='gray');
plt.axis('off');
plt.subplot(2,2,3)
plt.imshow(im_spiral_bw,cmap='gray');
plt.axis('off');
plt.subplot(2,2,4)
plt.imshow(im_amsterdam_bw,cmap='gray');
plt.axis('off');
```

### Create Sobel Filters

```python
Gx = np.zeros((3,3))
Gx[0,:] = [-1,+0,+1]
Gx[1,:] = [-2,+0,+2]
Gx[2,:] = [-1,+0,+1]

Gy = np.zeros((3,3))
Gy[0,:] = [+1,+2,+1]
Gy[1,:] = [+0,+0,+0]
Gy[2,:] = [-1,-2,-1]
```

### Convolution With Sobel Filters - Lenna

```python
imx = convolve2d(im_lenna_bw,Gx)
imy = convolve2d(im_lenna_bw,Gy)
imxy = np.sqrt(imx**2 + imy**2)
plt.figure(figsize=(10,10));
plt.subplot(2,2,1)
plt.imshow(im_lenna_bw,cmap='gray')
plt.axis('off');
plt.subplot(2,2,2)
plt.imshow(imx,cmap='gray')
plt.axis('off');
plt.subplot(2,2,3)
plt.imshow(imy,cmap='gray')
plt.axis('off');
plt.subplot(2,2,4)
plt.imshow(imxy,cmap='gray')
plt.axis('off');
```

### Convolution With Sobel Filters - Amsterdam

```python
imx = convolve2d(im_amsterdam_bw,Gx)
imy = convolve2d(im_amsterdam_bw,Gy)
imxy = np.sqrt(imx**2 + imy**2)
plt.figure(figsize=(10,10));
plt.subplot(2,2,1)
plt.imshow(im_amsterdam_bw,cmap='gray')
plt.axis('off');
plt.subplot(2,2,2)
plt.imshow(imx,cmap='gray')
plt.axis('off');
plt.subplot(2,2,3)
plt.imshow(imy,cmap='gray')
plt.axis('off');
plt.subplot(2,2,4)
plt.imshow(imxy,cmap='gray')
plt.axis('off');
```

### Gabor Patch Filter

```python
x = np.linspace(-10,10,1001)
X, Y = np.meshgrid(x, x)
wavelength = 16
angle = np.pi / 2
grating = np.sin(2*np.pi*(X*np.sin(angle+np.pi/4) + Y*np.cos(angle+np.pi/4)) / wavelength)

XX = X.flatten()
YY = Y.flatten()
XY = np.vstack((XX,YY)).T
gauss_distr = mnorm([0,0],[[15,0],[0,15]])
G = gauss_distr.pdf(XY)
G = np.reshape(G,((1001,1001)))
G[G < 0.0001] = 0
gabor = grating*G

plt.figure(figsize=(8,8));
plt.subplot(2,2,1);
plt.imshow(grating,cmap='gray');
plt.axis('off');
plt.subplot(2,2,2);
plt.imshow(gabor,cmap='gray');
plt.axis('off');
```

### Scale down to typical convolution node 'receptive field'

```python
gaborim = gabor
gaborim = gaborim - np.min(gaborim)
gaborim = gaborim / np.max(gaborim) * 255
gaborim = gaborim.astype('int8');
gaborim = Image.fromarray(gaborim);
gaborim = gaborim.resize((16,16))
G = np.array(gaborim)
plt.imshow(G,cmap='gray')
plt.axis('off');
```

### 2D Convolution

```python
imfilter1 = convolve2d(im_amsterdam_bw,G,'same');
imfilter2 = convolve2d(im_amsterdam_bw,G.T,'same');
plt.figure(figsize=(8,6))
plt.imshow(imfilter2-imfilter1,cmap='gray');
plt.axis('off');
```

```python
imfilter1 = convolve2d(im_spiral_bw,G,'same');
imfilter2 = convolve2d(im_spiral_bw,G.T,'same');
plt.figure(figsize=(8,6))
plt.imshow(imfilter1-imfilter2,cmap='gray');
plt.axis('off');
```
