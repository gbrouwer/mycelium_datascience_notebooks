### Read Categories

```python
codes = []
classnames = []
codedict = {}
classdict = {}
labeldict = {}
classcode = -1
with open('../../data/imagenet/meta/100_categories','r') as f:
    for line in f:
        elements = line.rstrip().split(':')
        code = elements[0]
        codes.append(code)
        classname = elements[1].replace("'","")
        classname = classname.split(',')[0][1:].lower()
        classnames.append(classname)
        if (classname not in labeldict):
            classcode = classcode + 1
            labeldict[classname] = classcode
        codedict[code] = classname
        classdict[classname] = code
```

### Find All Files and create y/ylabel

```python
myfiles = []
ylabels = []
y = []
for root, dirs, files in os.walk("../../data/imagenet/orig/train/", topdown=False):
    for name in files:
        if 'JPEG' in name:
            myfiles.append(os.path.join(root, name))
            code = root.split('/')[-1]
            classname = codedict[code]
            ylabels.append(classname)
            y.append(labeldict[classname])
myfiles = np.array(myfiles)
ylabels = np.array(ylabels)
y = np.array(y)
```

### Randomize

```python
randval = np.random.permutation(len(myfiles))
myfiles = myfiles[randval]
ylabels = ylabels[randval]
y = y[randval]
```

### Load and square all images

```python
fileno = 0
data = {}
indices = np.linspace(0,len(myfiles)-1,50).astype(int)
for j in range(len(indices)-1):
    si = indices[j]
    ei = indices[j+1]
    images_subset = []
    y_subset = []
    ylabels_subset = []
    for i in tqdm(range(si,ei)):
        imfile = myfiles[i]
        ylabels_subset.append(ylabels[i])
        y_subset.append(y[i])
        im = Image.open(imfile)
        width, height = im.size
        minext = np.min([width,height])
        left = np.ceil((width - minext)/2)
        top = np.ceil((height - minext)/2)
        right = np.ceil((width + minext)/2)
        bottom = np.ceil((height + minext)/2)
        im = im.crop((left, top, right, bottom))
        im = im.resize((240,240))
        im = np.array(im)
        if (len(im.shape) == 2):
            im = np.tile(im,(3,1,1))
            im = np.transpose(im,((1,2,0)))
        images_subset.append(im)
    data['y'] = np.array(y_subset)
    data['images'] = np.array(images_subset)
    data['y_labels'] = np.array(ylabels_subset)
    pickle.dump( data, open("../../data/imagenet/pickle/train/" + str(j) + ".p", "wb" ))
```

### Read back in

```python
data = pickle.load( open("../../data/imagenet/pickle/train/37.p", "rb"))
y = data['y']
y_labels = data['y_labels']
images = data['images']
```
