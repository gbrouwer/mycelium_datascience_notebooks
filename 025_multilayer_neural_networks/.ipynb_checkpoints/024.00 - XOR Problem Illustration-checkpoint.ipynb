### Import

```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from scipy.stats import multivariate_normal
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from mpl_toolkits import mplot3d
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Read Colormap

```python
cmap = Image.open('../../assets/colormaps/green_yellow_orange_red_muted.png')
colormap = np.array(cmap)[100,:,:]
```

### Create Data

```python
npoints = 50
X1 = multivariate_normal.rvs(mean=[+8,+8], cov=np.eye(2),size=int(npoints/2));
X2 = multivariate_normal.rvs(mean=[-8,-8], cov=np.eye(2),size=int(npoints/2));
X3 = multivariate_normal.rvs(mean=[-8,+8], cov=np.eye(2),size=int(npoints/2));
X4 = multivariate_normal.rvs(mean=[+8,-8], cov=np.eye(2),size=int(npoints/2));
X1c = X1.copy()
X2c = X2.copy()
X3c = X3.copy()
X4c = X4.copy()
X1c[:,1] = X1c[:,1] - 16
X1c[:,0] = X1c[:,0] - 16
X3c[:,0] = X3c[:,0] + 16
X4c[:,1] = X4c[:,1] +  16
Xs = np.vstack((X4,X2,X3,X1))
Xn = np.vstack((X1,X2,X3,X4))
Xc = np.vstack((X1c,X2c,X3c,X4c))
y = np.vstack((np.zeros((npoints,1)),np.ones((npoints,1))))[:,0]
```

### Meshgrid to visualize the boundaries

```python
vx,vy = np.meshgrid(np.linspace(-15,15,150),np.linspace(-15,15,150))
xx = vx.flatten()
yy = vy.flatten()
MX = np.vstack((xx,yy)).T
```

### Divide the data

```python
class0_ind = np.where(y == 0)[0]
class1_ind = np.where(y == 1)[0]
```

### Visualize

```python
plt.figure(figsize=(8,4));
plt.subplot(1,2,1);
plt.subplots_adjust(left=0.0,bottom=0.1, right=1.0, top=0.9, wspace=0.4, hspace=0.5)
plt.scatter(Xs[class0_ind,0],Xs[class0_ind,1],s=400,marker='^',alpha=0.5,facecolor=(0.00,0.70,0.45),edgecolor='k');
plt.scatter(Xs[class1_ind,0],Xs[class1_ind,1],s=400,marker='v',alpha=0.5,facecolor=(1.00,0.00,0.00),edgecolor='k');
plt.xticks(np.linspace(-12,12,5),np.linspace(-12,12,5),**csfont,fontsize=16)
plt.yticks(np.linspace(-12,12,5),np.linspace(-12,12,5),**csfont,fontsize=16)
plt.xlabel('Feature 1',**csfont,fontsize=24)
plt.ylabel('Feature 2',**csfont,fontsize=24)
plt.axis((-12,12,-12,12))

plt.subplot(1,2,2)
plt.scatter(Xn[class0_ind,0],Xn[class0_ind,1],s=400,marker='^',alpha=0.5,facecolor=(0.00,0.70,0.45),edgecolor='k');
plt.scatter(Xn[class1_ind,0],Xn[class1_ind,1],s=400,marker='v',alpha=0.5,facecolor=(1.00,0.00,0.00),edgecolor='k');
plt.xticks(np.linspace(-12,12,5),np.linspace(-12,12,5),**csfont,fontsize=16)
plt.yticks(np.linspace(-12,12,5),np.linspace(-12,12,5),**csfont,fontsize=16)
plt.xlabel('Feature 1',**csfont,fontsize=24)
plt.ylabel('Feature 2',**csfont,fontsize=24)
plt.axis((-12,12,-12,12));
plt.savefig('../../assets/matplotlib/024_data_sep_nosep.png', dpi=300,bbox_inches="tight")
```

### Can we find a transform for A -> C and for A -> B?

```python
Xsi = np.hstack((Xs,np.ones((Xs.shape[0],1))))
Xni = np.hstack((Xn,np.ones((Xn.shape[0],1))))
Xci = np.hstack((Xc,np.ones((Xc.shape[0],1))))
wsc = np.dot(np.dot(np.linalg.inv(np.dot(Xsi.T,Xsi)),Xsi.T),Xci)
wnc = np.dot(np.dot(np.linalg.inv(np.dot(Xni.T,Xni)),Xni.T),Xci)
Xsi_transformed = np.dot(Xsi,wsc)
Xni_transformed = np.dot(Xni,wnc)
Xsi_transformed[:,1] = 0
Xni_transformed[:,1] = 0

plt.figure(figsize=(8,10))
plt.subplots_adjust(left=0.0,bottom=0.1, right=1.0, top=0.9, wspace=0.5, hspace=0.5)

ax = plt.subplot(2,2,1)
ax.set_axisbelow(True);
plt.scatter(Xsi[class0_ind,0],Xsi[class0_ind,1],s=400,marker='^',alpha=0.5,facecolor=(0.00,0.70,0.45),edgecolor='k');
plt.scatter(Xsi[class1_ind,0],Xsi[class1_ind,1],s=400,marker='v',alpha=0.5,facecolor=(1.00,0.00,0.00),edgecolor='k');
plt.xticks(np.linspace(-12,12,5),np.linspace(-12,12,5),**csfont,fontsize=16)
plt.yticks(np.linspace(-12,12,5),np.linspace(-12,12,5),**csfont,fontsize=16)
plt.xlabel('Feature 1',**csfont,fontsize=24)
plt.ylabel('Feature 2',**csfont,fontsize=24)

plt.subplot(2,2,2)
plt.scatter(Xsi_transformed[class0_ind,0],Xsi_transformed[class0_ind,1],s=400,marker='^',alpha=0.5,facecolor=(0.00,0.70,0.45),edgecolor='k');
plt.scatter(Xsi_transformed[class1_ind,0],Xsi_transformed[class1_ind,1],s=400,marker='v',alpha=0.5,facecolor=(1.00,0.00,0.00),edgecolor='k');
plt.xticks(np.linspace(-12,12,5),np.linspace(-12,12,5),**csfont,fontsize=16)
plt.yticks(np.linspace(-0.5,0.5,5),np.linspace(-0.5,0.5,5),**csfont,fontsize=16)
plt.xlabel('Feature 1',**csfont,fontsize=24)
plt.ylabel('Feature 2',**csfont,fontsize=24)

plt.subplot(2,2,3)
plt.subplots_adjust(left=0.0,bottom=0.1, right=1.0, top=0.9, wspace=0.5, hspace=0.5)
plt.scatter(Xni[class0_ind,0],Xni[class0_ind,1],s=400,marker='^',alpha=0.5,facecolor=(0.00,0.70,0.45),edgecolor='k');
plt.scatter(Xni[class1_ind,0],Xni[class1_ind,1],s=400,marker='v',alpha=0.5,facecolor=(1.00,0.00,0.00),edgecolor='k');
plt.xticks(np.linspace(-12,12,5),np.linspace(-12,12,5),**csfont,fontsize=16)
plt.yticks(np.linspace(-12,12,5),np.linspace(-12,12,5),**csfont,fontsize=16)
plt.xlabel('Feature 1',**csfont,fontsize=24)
plt.ylabel('Feature 2',**csfont,fontsize=24)


plt.subplot(2,2,4)
plt.scatter(Xni_transformed[class0_ind,0],Xni_transformed[class0_ind,1],s=400,marker='^',alpha=0.5,facecolor=(0.00,0.70,0.45),edgecolor='k');
plt.scatter(Xni_transformed[class1_ind,0],Xni_transformed[class1_ind,1],s=400,marker='v',alpha=0.5,facecolor=(1.00,0.00,0.00),edgecolor='k');
plt.xticks(np.linspace(-0.5,0.5,5),np.linspace(-0.5,0.5,5),**csfont,fontsize=16)
plt.yticks(np.linspace(-0.5,0.5,5),np.linspace(-0.5,0.5,5),**csfont,fontsize=16)
plt.ylabel('Feature 2',**csfont,fontsize=24)
plt.xlabel('Feature 1',**csfont,fontsize=24)
plt.savefig('../../assets/matplotlib/024_data_transformed.png', dpi=300,bbox_inches="tight")
```

### Divide Training and Testing Data

```python
Xs_train, Xs_test, ys_train, ys_test = train_test_split(Xs, y, test_size=0.20, random_state=42)
Xns_train, Xns_test, yns_train, yns_test = train_test_split(Xn, y, test_size=0.20, random_state=42)
```

### Create Perceptron

```python
perceptron_sep = Perceptron(max_iter=20000)
perceptron_nosep = Perceptron(max_iter=20000)
```

### Train on Seperable and Non-Seperable Data

```python
perceptron_sep.fit(Xs_train, ys_train);
perceptron_nosep.fit(Xns_train, yns_train);
print(perceptron_sep.score(Xs_test,ys_test))
print(perceptron_nosep.score(Xns_test,yns_test))
```

### Get Decision Map for Seperable Data

```python
#Get Weights and confidence
Confs = perceptron_sep.decision_function(MX)
ws = perceptron_sep.coef_
bs = perceptron_sep.intercept_
wn = perceptron_nosep.coef_
bn = perceptron_nosep.intercept_

#Decision Boundary
slope = ws[0][0]/ws[0][1]
xval = np.linspace(-15,15,100)
yval = xval*slope

#Compute activation and prediction
X_prime = np.hstack((MX,np.ones((MX.shape[0],1))))
wn_prime = np.vstack((wn[0,0],wn[0,1],bn))
an_prime = np.dot(X_prime,wn_prime)
yn_prime = ((an_prime > 0)*1.0)[:,0]
Confs = Confs - np.min(Confs)
Confs = np.floor(Confs / np.max(Confs) * 255)
Confs = Confs.astype(int)
cmap = colormap[Confs,:]
cmap = np.reshape(cmap,((150,150,3)));
Confs = np.reshape(Confs,(150,150))

#Plot
plt.figure(figsize=(12,6));
ax = plt.subplot(1,2,1);
ax.set_axisbelow(True);
plt.plot(xval,yval,'k--',linewidth=2);
plt.scatter(Xs[class0_ind,0],Xs[class0_ind,1],s=400,marker='^',alpha=0.5,facecolor=(0.00,0.70,0.45),edgecolor='k');
plt.scatter(Xs[class1_ind,0],Xs[class1_ind,1],s=400,marker='v',alpha=0.5,facecolor=(1.00,0.00,0.00),edgecolor='k');
plt.imshow(cmap,extent=(-15,15,-15,15));
plt.xticks(np.linspace(-15,15,7),np.linspace(-15,15,7),**csfont,fontsize=16);
plt.yticks(np.linspace(-15,15,7),np.linspace(-15,15,7),**csfont,fontsize=16);
plt.ylabel('Feature 2',**csfont,fontsize=24)
plt.xlabel('Feature 1',**csfont,fontsize=24)
plt.savefig('../../assets/matplotlib/024_perceptron_sep.png', dpi=300,bbox_inches="tight")
```

### Get Decision Map for Non-Seperable Data

```python
#Get Weights and confidence
Confn = perceptron_nosep.decision_function(MX)
wn = perceptron_nosep.coef_
bn = perceptron_nosep.intercept_

#Decision Boundary
slope = wn[0][0]/wn[0][1]
xval = np.linspace(-15,15,100)
yval = xval*slope

##compute activation and prediction
X_prime = np.hstack((MX,np.ones((MX.shape[0],1))))
wn_prime = np.vstack((wn[0,0],wn[0,1],bn))
an_prime = np.dot(X_prime,wn_prime)
yn_prime = ((an_prime > 0)*1.0)[:,0]
Confn = Confn - np.min(Confn)
Confn = np.floor(Confn / np.max(Confn) * 255)
Confn = Confn.astype(int)

rgb = np.zeros((256,3))
rgb[:,0] = np.linspace(0,1.0,256)
rgb[:,1] = 1.0 - np.linspace(0,1,256)
rgb[:,2] = 0.5
cmap = colormap[Confn,:]
cmap = np.reshape(cmap,((150,150,3)));

plt.figure(figsize=(12,6));
ax = plt.subplot(1,2,1);
plt.plot(xval,yval,'k--',linewidth=2);
ax.set_axisbelow(True);
plt.scatter(Xn[class0_ind,0],Xn[class0_ind,1],s=400,marker='^',alpha=0.5,facecolor=(0.00,0.70,0.45),edgecolor='k');
plt.scatter(Xn[class1_ind,0],Xn[class1_ind,1],s=400,marker='v',alpha=0.5,facecolor=(1.00,0.00,0.00),edgecolor='k');
plt.imshow(cmap,extent=(15,-15,15,-15));
plt.xticks(np.linspace(-15,15,7),np.linspace(-15,15,7),**csfont,fontsize=16);
plt.yticks(np.linspace(-15,15,7),np.linspace(-15,15,7),**csfont,fontsize=16);
plt.savefig('../../assets/matplotlib/024_perceptron_nosep.png', dpi=300,bbox_inches="tight")
plt.ylabel('Feature 2',**csfont,fontsize=24);
plt.xlabel('Feature 1',**csfont,fontsize=24);
```
