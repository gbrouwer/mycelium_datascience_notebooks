### Import

```python
import os
import numpy as np
import sys
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import spectral.io.aviris as aviris
import colour
from PIL import Image
from scipy.stats import multivariate_normal as mnorm
from scipy import ndimage
from matplotlib import cm
from mpl_toolkits import mplot3d
from skimage import io, color
from matplotlib.collections import PatchCollection
from matplotlib.patches import Rectangle
import matplotlib.tri as mtri
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Read Painting

```python
painting = Image.open("../../assets/images/paintings/giorgio_de_chirico_piazza_ditalia.jpg")
painting = np.array(painting)
```

### Reshape and Convert to LAB

```python
RGB = np.reshape(painting,((painting.shape[0]*painting.shape[1],painting.shape[2])))
LAB = color.rgb2lab(RGB)
L = np.reshape(LAB,((painting.shape[0],painting.shape[1],painting.shape[2])))
L = L[:,:,0]
```

### Original

```python
plt.figure(figsize=(8,8))
plt.imshow(painting);
plt.axis('off');
plt.savefig('../../assets/matplotlib/019_painting_orig.png',dpi=300,bbox_inches="tight")
```

### Plot Luminance Channel

```python
plt.figure(figsize=(8,8))
plt.imshow(L[:,:],cmap='gray');
plt.axis('off');
plt.savefig('../../assets/matplotlib/019_painting_luminance.png',dpi=300,bbox_inches="tight")
```

### Create and Plot Saturation Channel

```python
D = np.sqrt(LAB[:,1]**2 + LAB[:,2]**2)
D = np.reshape(D,(painting.shape[0],painting.shape[1]))
plt.figure(figsize=(8,8))
plt.imshow(D[:,:],cmap='gray');
plt.axis('off');
plt.savefig('../../assets/matplotlib/019_painting_saturation.png',dpi=300,bbox_inches="tight")
```

### Draw Isoluminant Image

```python
Lmean = np.mean(LAB[:,0])
LABiso = LAB.copy()
LABiso[:,0] = 50
RGBiso = color.lab2rgb(LABiso)
RGBiso = np.reshape(RGBiso,(painting.shape[0],painting.shape[1],painting.shape[2]))
plt.figure(figsize=(8,8))
plt.imshow(RGBiso)
plt.axis('off');
plt.savefig('../../assets/matplotlib/019_painting_isolum.png',dpi=300,bbox_inches="tight")
```
