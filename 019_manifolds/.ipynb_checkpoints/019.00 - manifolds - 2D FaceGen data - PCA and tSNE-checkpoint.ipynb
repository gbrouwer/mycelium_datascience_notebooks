### Parameters

```python
no_components = 16
imsize = 256
dictionary_name = 'skin_textures_' + str(imsize)
```

### Combine skin textures into large matrix XT

```python
Xsk = []
for i in range(skin_textures.shape[0]):
    xsk = skin_textures[i,:,:,:]
    xsk = np.reshape(xsk,(xsk.shape[0]*xsk.shape[1]*xsk.shape[2],1))
    Xsk.append(xsk)
Xsk = np.squeeze(np.array(Xsk))
skin_textures_mean = np.mean(skin_textures,axis=0)
skin_textures_mean = skin_textures_mean / 255
skin_textures_mean_vector = np.reshape(skin_textures_mean,(imsize*imsize,3))
plt.imshow(skin_textures_mean);
plt.axis('off');
```

### Setup and Run PCA

```python
pca = PCA(n_components=no_components)
C = pca.fit_transform(Xsk.T)
C = C.T
```

### Log of Explained variance 

```python
xticks = np.linspace(1,no_components,no_components)
yticks = np.linspace(0.0005,1,12,True)
ytickslabels = np.round(yticks,2).astype('str')
ytickslabels[::2] = ""
plt.figure(figsize=(8,8))
ax = plt.subplot();
plt.plot(range(1,no_components+1),pca.explained_variance_ratio_,c='k');
plt.yscale('log');
plt.grid();
plt.scatter(xticks,pca.explained_variance_ratio_,c='r',edgecolor='k',s=200,zorder=100);
plt.xticks(xticks,labels=xticks.astype(int),**csfont,fontsize=18);
plt.yticks(yticks,labels=ytickslabels,rotation=0,**csfont,fontsize=18);
plt.axis((0,no_components+1,0.0005,1.1));
plt.ylabel('Explained Variance',**csfont,fontsize=24);
plt.xlabel('Component',**csfont,fontsize=24);
plt.savefig('../../assets/matplotlib/019_faces_texture_explained_variance.png',dpi=300,bbox_inches="tight")
```

### Normalize the components to RGB by finding the transform of the first component (representing the mean) and the actual mean image

```python
B = np.ones((imsize*imsize,1))
C0 = C[0,:]
C0 = np.reshape(C0,(imsize*imsize,3))
C0 = np.hstack((C0,B))
MV = np.hstack((skin_textures_mean_vector,B))
T = np.dot(np.dot(np.linalg.inv(np.dot(C0.T,C0)),C0.T),MV)
```

### Plot the first 10 components in RGB space

```python
plt.figure(figsize=(12,12))
B = np.ones((imsize*imsize,1))
for i in range(no_components):
    Cc = C[i,:]
    Cc = np.reshape(Cc,(imsize*imsize,3))
    if (i > 0): Cc = Cc * 6
    Cc = np.hstack((Cc,B))
    Crgb = np.dot(Cc,T)[:,:-1]
    Crgb = np.reshape(Crgb,(imsize,imsize,3))
    Crgb[Crgb > 1] = 1
    Crgb[Crgb < 0] = 0
    ax = plt.subplot(4,4,i+1);
    plt.imshow(Crgb)
    plt.axis('off')
plt.savefig('../../assets/matplotlib/019_faces_texture_first_16_components.png',dpi=300,bbox_inches="tight")    
```

### Create a large number of random images by indepdently varying how much each component after the mean is added

```python
eig = pca.explained_variance_ratio_[:5]
eig = eig[1] / eig 
eig[0] = 1
eig = np.expand_dims(eig,-1)
Xtext = []
weights = []
for i in range(10):
    w = (np.random.random((5,1)) - 0.5) * 2
    w = w * eig
    w[0] = 1
    G = np.dot(C[:5,:].T,w)
    G = np.reshape(G,(imsize*imsize,3))
    G = np.hstack((G,B))
    Grgb = np.dot(G,T)[:,:-1]
    Grgb = np.reshape(Grgb,(imsize,imsize,3))
    Grgb[Grgb > 1] = 1
    Grgb[Grgb < 0] = 0
    Grgb = Grgb[::4,::4,:]
    Grgb = np.reshape(Grgb,(Grgb.shape[0]*Grgb.shape[1]*Grgb.shape[2],1))
```

### Create PCA (linear) Manifold
Create a 11 by 11 grids of two components varying from positive to negative for the first 5 components after component 0 (which represents the mean of all face textures)

```python
res = 3
C0 = C[0,:]
C0 = np.reshape(C0,(imsize*imsize,3))
B = np.ones((C0.shape[0],1))
cx = np.linspace(-2,2,res)
cy = np.linspace(-2,2,res)
for m in range(1,3):
    for l in range(m+1,3):
        Cx = C[m,:]
        Cx = np.reshape(Cx,(imsize*imsize,3))
        Cy = C[l,:]
        Cy = np.reshape(Cy,(imsize*imsize,3))
        for i in range(cx.shape[0]):
            for j in range(cy.shape[0]):
                
                #Create Weighted Texture
                valx = cx[i]
                valy = cy[j]
                Ccombined = C0 + Cx*valx + Cy*valy
                Ccombined = np.hstack((Ccombined,B))
                Crgb = np.dot(Ccombined,T)[:,:-1]
                Crgb = np.reshape(Crgb,(imsize,imsize,3))
                Crgb[Crgb > 1] = 1
                Crgb[Crgb < 0] = 0

                #Write texture file
                texture_savename = '../../data/facegen/2d_texture_space/c' + str(m) + '_c' + str(l) + '_' + str(i) + '_' + str(j) + '_skin_hi.jpg'
                im = Image.fromarray(np.uint8(Crgb*255))
                im = im.save(texture_savename)
                
                #Write MTL file
                mtl_texture_ref = 'c' + str(m) + '_c' + str(l) + '_' + str(i) + '_' + str(j) + '_skin_hi.jpg'
                mtl_file[mtl_index] = '    map_Kd ' + mtl_texture_ref
                mtl_savename = '../../data/facegen/2d_texture_space/c' + str(m) + '_c' + str(l) + '_' + str(i) + '_' + str(j) + '.mtl'
                with open(mtl_savename,'w') as f:
                    for line in mtl_file:
                        f.write(line + '\n')
                
                #Write OBJ file
                count = 0
                indexer = 1
                obj_indices = np.array(obj_indices)
                obj_file[4] = 'mtllib ' + mtl_ref
                obj_savename = '../../data/facegen/2d_texture_space/c' + str(m) + '_c' + str(l) + '_' + str(i) + '_' + str(j) + '.obj'
                with open(obj_savename,'w') as f:
                    for myline in obj_file:
#                         if ('v ' in myline[:2]):
#                             new_vector = np.round(CA[indexer,:],4).astype(str)
#                             new_vector_str = 'v ' + ' '.join(new_vector)
#                             f.write(new_vector_str + '\n')
#                             indexer = indexer + 1
#                         else:
                        f.write(myline + '\n')
                        count = count + 1
```
