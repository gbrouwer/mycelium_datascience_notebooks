### Load LMS Curves

```python
lms = pd.read_csv('../../data/color/lms_adorption_spectra/lms-curves.csv',delimiter=',',header=None)
lms = lms.fillna(0)
lms = lms.values
idx = np.where(lms[:,0] < 700)[0]
lms = lms[idx,:]
idx = np.where(lms[:,0] > 400)[0]
lms = lms[idx,:]
lcone_idx = np.where(lms[:,1] > 0.02)[0]
mcone_idx = np.where(lms[:,2] > 0.02)[0]
scone_idx = np.where(lms[:,3] > 0.02)[0]
maxval = np.argmax(lms,axis=0)
lmax = round(lms[maxval[1],0])
mmax = round(lms[maxval[2],0])
smax = round(lms[maxval[3],0])
print(lmax)
print(mmax)
print(smax)
```

### Print Visibile Spectrum and Cone Sensititives

```python
spectrum_image = Image.open("../../assets/figures/019/spectrum.png")
spectrum_image = np.array(spectrum_image)
plt.figure(figsize=(12,9))
plt.imshow(spectrum_image,extent=(400,700,-0.2,1.4),aspect=100)
plt.plot(lms[lcone_idx,0],lms[lcone_idx,1],c='w',linewidth=3)
plt.plot(lms[mcone_idx,0],lms[mcone_idx,2],c='w',linewidth=3)
plt.plot(lms[scone_idx,0],lms[scone_idx,3],c='w',linewidth=3)
plt.grid();
plt.yticks([]);
plt.xticks(fontsize=14,**csfont)
plt.xticks(np.linspace(400,700,7,True),labels=np.floor(np.linspace(400,700,7,True)).astype(int),fontsize=14,**csfont);
plt.axis((400,700,-0.2,1.4))
plt.arrow(442,1.20,0,-0.10,color='w',
          width=1,
          length_includes_head=True,
          head_width=5,
          head_length=0.025,
          overhang=0,
          head_starts_at_zero=True)
plt.arrow(542,1.20,0,-0.10,color='w',
          width=1,
          length_includes_head=True,
          head_width=5,
          head_length=0.025,
          overhang=0,
          head_starts_at_zero=True)
plt.arrow(569,1.20,0,-0.10,color='w',
          width=1,
          length_includes_head=True,
          head_width=5,
          head_length=0.025,
          overhang=0,
          head_starts_at_zero=True)
plt.text(439,1.25,"S",c='w',fontsize=20,**csfont,fontweight='bold')
plt.text(538,1.25,"M",c='w',fontsize=20,**csfont,fontweight='bold')
plt.text(565,1.25,"L",c='w',fontsize=20,**csfont,fontweight='bold')
plt.grid();
plt.xlabel('Wavelength (nanometers)',fontsize=18,**csfont,labelpad=12);
plt.ylabel('Sensitivity (arbitrary units)',fontsize=18,**csfont,labelpad=12);
plt.savefig('../../assets/figures/019/spectrum1.png', dpi=300,bbox_inches="tight")
```
