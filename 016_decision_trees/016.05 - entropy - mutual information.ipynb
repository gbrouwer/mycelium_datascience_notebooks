### Entropy - Biased Coin Example
The concept of entropy in information theory is perhaps best explained using a coin flip. Imagine a coin that is completely biased and no matter what we do, it will always end up heads (or tails). If you knew this to be true, flipping the coin and seeing its usual outcome would not come as a suprise as all. This lack of surprise is associated with low entropy. If, however, the coin is completely unbiased, we have no way of predicting the outcome with any certainty. In other words, the outcome of the coin flip will always be a suprise, which is associated with high entropy. This parallels the notion of entropy in physics: a system in perfect order holds no surprises, observing some of it gives us a good insight of the entire system. However, a system in disorder cannot be described easiy. Randomness in both physical systems and data leads to surprises: you really never know what comes next. Entropy is usually expressed in bits. Since a coin flip has two possible outcomes or states, the most information we can gain from observing the outcome of it is 1 bit, as 1 bit can encode two states. If the outcome is biased, we gain less information, and the entropy will be a fraction of the maximum entropy of 1. If the coin is biased completely, there is no information to be gained, and the entropy will be zero. 

```python
entropies = []
bias = []
for i in range(91):
    p = (i/100) + 0.05
    bias.append(p)
    H = -((p * np.log2(p)) + ((1-p) * np.log2(1-p)))
    entropies.append(H)
plt.figure(figsize=(5,4))
plt.plot(bias,entropies,'k');
plt.grid()
plt.xlabel('coin bias (probability)',fontsize=18)
plt.ylabel('entropy (bits)',fontsize=18);
plt.savefig('../../assets/figures/15_entropy.png');
```

### Entropy - Letters in the English Alphabet
A more interesting example of entropy is the relative frequency of the letters in the alphabet occuring in a corpus. Let us imagine we know nothing about english. This means that we can't possible know that some letters appear more frequently than others. So, observing the any letter has a probability of 1/26. If we compute the entropy of observing any letter in this scenario, the amount of entropy (information gained) is 4.7 bits. However, if we take into account that letters do not appear with the same frequency in english, we obtain an entropy of 4.18. 

```python
probs = {}
probs['E'] = 12.02
probs['T'] = 9.10
probs['A'] = 8.12
probs['O'] = 7.68
probs['I'] = 7.31
probs['N'] = 6.95
probs['S'] = 6.28
probs['R'] = 6.02
probs['H'] = 5.92
probs['D'] = 4.32
probs['L'] = 3.98
probs['U'] = 2.88
probs['C'] = 2.71
probs['M'] = 2.61
probs['F'] = 2.30
probs['Y'] = 2.11
probs['W'] = 2.09
probs['G'] = 2.03
probs['P'] = 1.82
probs['B'] = 1.49
probs['V'] = 1.11
probs['K'] = 0.69
probs['X'] = 0.17
probs['Q'] = 0.11
probs['J'] = 0.10
probs['Z'] = 0.07
for item in probs:
    probs[item] = probs[item] / 100.0
equal = 0
observed = 0
for letter in probs:
    equal = equal + ((1/26) * np.log2(1/26))
    observed = observed + (probs[letter]) * np.log2(probs[letter])
equal = -equal
observed = -observed
print(equal)
print(observed)
```

### Create Joint Distribution with and without covariance
Mutual information is a measurement of how the probability of observing one variable (X) is a function of another variable (Y). If knowledge of variable Y predicts the variable X, there is mutual information: information that is shared between them.

```python
mu_1 = [100,100]
mu_2 = [100,100]
sigma_1 = [[320,160],[160,160]]
sigma_2 = [[160,0],[0,320]]
rv_1 = mnorm(mu_1,sigma_1)
rv_2 = mnorm(mu_2,sigma_2)

x = np.linspace(50,150,80)
y = np.linspace(50,150,80)
X,Y = np.meshgrid(x, y)
XX = np.array([X.ravel(), Y.ravel()]).T
Z1 = rv_1.pdf(XX)
Z2 = rv_2.pdf(XX)
Z1 = Z1.reshape(X.shape)
Z2 = Z2.reshape(X.shape)

plt.figure(figsize=(10,4));
plt.subplot(1,2,1);
plt.contourf(Z1,levels=15,cmap='YlGnBu');
plt.contour(Z1,levels=15,colors='gray');
plt.subplot(1,2,2);
plt.contourf(Z2,levels=10,cmap='YlGnBu');
plt.contour(Z2,levels=10,colors='gray');
plt.savefig('../../assets/figures/15_mutual_information.png');
plt.show();
```

### Compute Mutual Information for both distributions

```python
pXY = Z1
pX = np.sum(Z1,axis=0)
pY = np.sum(Z1,axis=1)
pX = np.expand_dims(pX,axis=-1)
pY = np.expand_dims(pY,axis=-1)
I = np.sum(np.log2(pXY / np.dot(pX,pY.T)) * pXY)
print(I)

pXY = Z2
pX = np.sum(Z2,axis=0)
pY = np.sum(Z2,axis=1)
pX = np.expand_dims(pX,axis=-1)
pY = np.expand_dims(pY,axis=-1)
I = np.sum(np.log2(pXY / np.dot(pX,pY.T)) * pXY)
print(I)
```
