### PCA manifolds
Using PCA we can find a lower dimensional that accounts for as much of variance as possible. PCA is strictly linear and finds principal componenets orthogonal to each other. This makes it fast, but it will not be able to map the non-linear structures. Because of this, when we apply PCA, we are trying to find a hyperplane: a flat surface oriented within our data space. In the context of manifold learning, a hyperplane can be considered a manifold, just not a very interesting one, unable to account for non linearities by means of curvature. 

### Fit and Transform

```python
pca = PCA(n_components=100)
T = pca.fit_transform(Xtrain)
```

### Cumulative Explained Variance
The PCA was able to explain much of the data with smaller number of principal components, relative to the input dimensionality of 784. However, it still required a substantial amount of principal components (40) to get to around 80% explained variance, and further ~100 components to get to 90%. 

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_));
plt.xlabel('principal components');
plt.ylabel('cumulative explained variance');
```

### T-SNE
t-distributed stochastic neighbor embedding or T-SNE is a non linear approach for finding lower dimensional manifolds that capture the smoothly varying nature of our input data. It is computationally heavy, compared to linear methods like PCA, but can yield some very interesting results. Because of this, we will subsample the data to get a reasonable result, in a reasonable amount of time.

```python
subsample_factor = 25
tsne = TSNE(n_components=2, init="random", learning_rate="auto", random_state=0)
T = tsne.fit_transform(Xtrain[::subsample_factor,:], ytrain[::subsample_factor]);
yviz = ytrain[::subsample_factor]
```

### Lets visualize the results

```python
fig, ax = plt.subplots(figsize=(9, 9))
for i in range(10):
    indices = np.where(yviz == i)[0]
    ax.scatter(T[indices,0],T[indices,1],label=i)
ax.legend();
ax.grid();
plt.xlabel('PC2');
plt.ylabel('PC3');
```
