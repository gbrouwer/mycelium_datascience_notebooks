### Import

```python
import os
import sys
import numpy as np
import pandas as pd
import json
import networkx as nx
import matplotlib.pyplot as plt
from scipy.spatial.distance import squareform
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
from tqdm import tqdm
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Read City Data

```python
with open('../../data/us_cities/us_cities.json','r') as f:
    data = json.load(f)
```

### Create Point Cloud

```python
X = []
cities = []
for item in data:
    lat = item['latitude']
    lng = item['longitude']
    X.append((lng,lat))
    cities.append(item['city'])
X = np.array(X)
```

### Create Graph

```python
G = nx.Graph()
edges = []
max_neighbors = 6
D = squareform(pdist(X))
for i in range(D.shape[0]):
    weights = np.sort(D[:,i])[1:max_neighbors+1]
    indices = np.argsort(D[:,i])[1:max_neighbors+1]
    for j in range(len(indices)):
        edges.append((i,indices[j]))
        G.add_edge(i, indices[j], weight=1/weights[j])
edges = np.array(edges)
```

### Dijkstra for connection between LA and New York

```python
path = nx.shortest_path(G, source=0, target=1,weight='weight')
```

### Visualize

```python
plt.figure(figsize=(12,8))
plt.scatter(X[:,0],X[:,1],color=(0.3,0.6,0.9))
for i in range(len(path)-1):
    ind1 = path[i]
    ind2 = path[i+1]
    x1 = X[ind1,0]
    y1 = X[ind1,1]    
    x2 = X[ind2,0]
    y2 = X[ind2,1]        
    plt.plot([x1,x2],[y1,y2],c='k')
plt.grid()
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel('Longitude',fontsize=20,**csfont);
plt.ylabel('Latitude',fontsize=20,**csfont);
plt.title('Dijkstra\n',fontsize=24,**csfont);
plt.savefig('../../assets/figures/020/dijkstra.png',dpi=300,bbox_inches="tight")       
```
