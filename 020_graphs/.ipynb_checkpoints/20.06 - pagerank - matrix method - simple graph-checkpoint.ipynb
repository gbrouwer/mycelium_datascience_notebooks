# 20.06 - PageRank Matrix Method
This noteboom uses matrix multiplication to compute the PageRank values. This create a cleaner, more concise algorithm and code base implementing PageRank. 
More importantly, optimized toolboxes and hardware can dramatically speed up the computation for large graphs using this algebriac approach

### Import

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Define our graph to test our algorithm

```python
graph = [
    (2,1),
    (2,3),
    (2,5),
    (2,4),
    (2,7),
    (1,2),
    (3,2),
    (3,1),
    (5,1),
    (5,3),
    (5,4),
    (4,2),
    (4,3),
    (4,5),
    (4,6),
    (6,2),
    (6,4),
    (7,4)]
```

### Coordinates for plotting purposes

```python
coords = np.array([
    [-1.5,-0.5],
    [-0.5,0],
    [0.75,1],
    [0.25,0.25],
    [0.25,-0.5],
    [-0.75,0.5],
    [-0.25,1]])
coords[:,0] = coords[:,0] - np.mean(coords[:,0])
coords[:,1] = coords[:,1] - np.mean(coords[:,1])
```

### Find the unique number of nodes

```python
nodes = np.unique(np.array(graph))
n = len(nodes)
```

### Create Adjacency Matrix

```python
A = np.zeros((n,n))
for edge in graph:
    parent,child = edge
    A[parent-1,child-1] = 1
A = A.T
for i in range(A.shape[0]):
    A[:,i] = A[:,i] / np.sum(A[:,i]) 
```

### Successive Matrix Squaring

```python
no_iter = 20
for i in range(no_iter):
    A = np.dot(A,A)
v = np.ones((n,1)) / n
pageranks = np.dot(A,v)
print(pageranks)
```

### Plot the resulting PageRanks on the graph

```python
plt.figure(figsize=(8,8))
norm_pageranks = pageranks / np.max(pageranks)
for i in range(len(nodes)):
    pr = pageranks[i,0]
    norm_pr = norm_pageranks[i,0]
    pr = np.round(pr,2)
    plt.scatter(coords[i,0],coords[i,1],s=1000,
                color=(norm_pr,0,1-(norm_pr**2)),
                zorder=100)
    plt.text(coords[i,0]-0.05,coords[i,1]-0.05,str(i+1),
             color='white',
             zorder=101,fontsize=18,**csfont)
    plt.text(coords[i,0]-0.30,coords[i,1]+0.15,str(pr),
             color='black',
             zorder=101,fontsize=18,**csfont)

plt.grid()
plt.axis((-1.5,1.5,-1.5,1.5))
for edge in graph:
    left,right = edge
    plt.plot([coords[left-1,0],coords[right-1,0]],[coords[left-1,1],coords[right-1,1]],'k')
plt.xlabel('x',fontsize=18,**csfont);
plt.ylabel('y',fontsize=18,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.savefig('../../assets/figures/020/pagerank_4.png',dpi=300,bbox_inches="tight")    
```
