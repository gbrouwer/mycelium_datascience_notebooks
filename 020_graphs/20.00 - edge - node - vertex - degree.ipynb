# 20.00 - Graph Essential Terminology
This script generates the figure highlighting the key terms in graph theory

### Import

```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Fonts for display

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Create Three Nodes

```python
head_width = 0.25
head_length = 0.25
edge_width = 0.05
X = np.array([[0,2],[-2,-2],[2,-2]])
plt.figure(figsize=(6,6))
plt.scatter(X[:,0],X[:,1],s=4000,linewidth=2,facecolor='w',edgecolor='k',zorder=1000)
plt.axis((-4,4,-4,4));
plt.axis('off');
plt.arrow(-0.25,2,-1.6,-3.25,width=edge_width,
          length_includes_head=True,
          head_width=head_width,
          head_length=head_length,facecolor='k',edgecolor='k')
plt.arrow(-1.75,-2,1.6,3.2,width=edge_width,
          length_includes_head=True,
          head_width=head_width,
          head_length=head_length,facecolor='k',edgecolor='k')
plt.arrow(0.25,2,1.6,-3.25,width=edge_width,
          length_includes_head=True,
          head_width=head_width,
          head_length=head_length,facecolor='k',edgecolor='k')
plt.arrow(1.75,-2,-1.6,3.2,width=edge_width,
          length_includes_head=True,
          head_width=head_width,
          head_length=head_length,facecolor='k',edgecolor='k')
plt.arrow(-2,-2,3.25,0,width=edge_width,
          length_includes_head=True,
          head_width=head_width,
          head_length=head_length,facecolor='k',edgecolor='k')
plt.text(x=-0.15,y=1.85,s='A',zorder=2000,**csfont,fontsize=24);
plt.text(x=-2.15,y=-2.15,s='B',zorder=2000,**csfont,fontsize=24);
plt.text(x=+1.85,y=-2.15,s='C',zorder=2000,**csfont,fontsize=24);
plt.text(x=+3.75,y=-2.15,s='Vertex or Node',zorder=2000,**csfont,fontsize=24);
plt.arrow(3.5,-2,-0.5,0,width=edge_width*0.5,
          length_includes_head=True,
          head_width=head_width,
          head_length=head_length,facecolor='k',edgecolor='k');
plt.text(x=2.25,y=0.15,s='Edge',zorder=2000,**csfont,fontsize=24);
plt.arrow(2,+0.25,-0.5,0,width=edge_width*0.5,
          length_includes_head=True,
          head_width=head_width,
          head_length=head_length,facecolor='k',edgecolor='k');
plt.text(x=1.5,y=2.50,s='Degree of A is 4',zorder=2000,**csfont,fontsize=16);
plt.text(x=1.5,y=2.00,s='In Degree of A is 2',zorder=2000,**csfont,fontsize=16);
plt.text(x=1.5,y=1.50,s='Out Degree of A is 2',zorder=2000,**csfont,fontsize=16);
plt.arrow(1.4,2.1,-0.5,0,width=edge_width*0.5,
          length_includes_head=True,
          head_width=head_width,
          head_length=head_length,facecolor='k',edgecolor='k');
plt.savefig('../../assets/matplotlib/020_graph_essentials.png',dpi=300,bbox_inches="tight")
```
