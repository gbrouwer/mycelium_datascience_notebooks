# Activation Functions
Some plots to visualize activation function

### Import

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Sigmoid function

```python
def sigmoid(x):
    return 1  / (1 + np.exp(-x))
```

### Linear function

```python
plt.figure(figsize=(6,6));
x = np.linspace(-2,2,101)
y = x.copy()
plt.plot(x,y,'k')
plt.xticks(np.linspace(-2,2,11),labels=np.round(np.linspace(-2,2,11),2),fontsize=16,**csfont);
plt.yticks(np.linspace(-2,2,11),labels=np.round(np.linspace(-2,2,11),2),fontsize=16,**csfont);
plt.xlabel('input',fontsize=24,**csfont,labelpad=20);
plt.ylabel('output',fontsize=24,**csfont);
plt.text(-3,2.45,'(A)',fontsize=32,**csfont);
plt.title('Linear\n',fontsize=24,**csfont);
plt.grid()
plt.savefig('../../assets/matplotlib//022_activation_function_1.png', dpi=300,bbox_inches="tight")
```

### Rectified Linear - ReLu

```python
plt.figure(figsize=(6,6));
x = np.linspace(-2,2,101)
y = x.copy()
y[:51] = 0
y = y / 2
plt.plot(x,y,'k')
plt.xticks(np.linspace(-2,2,11),labels=np.round(np.linspace(-2,2,11),2),fontsize=16,**csfont);
plt.yticks(np.linspace(0,1,11),labels=np.round(np.linspace(0,1,11),2),fontsize=16,**csfont);
plt.xlabel('input',fontsize=24,**csfont,labelpad=20);
plt.ylabel('output',fontsize=24,**csfont);
plt.text(-3,1.12,'(D)',fontsize=32,**csfont);
plt.title('ReLU\n',fontsize=22,**csfont);
plt.grid()
plt.savefig('../../assets/matplotlib//022_activation_function_2.png', dpi=300,bbox_inches="tight")
```

### Leaky ReLU

```python
plt.figure(figsize=(6,6));
x = np.linspace(-2,2,101)
y = x.copy()
y[:51] = y[:51] * 0.5
y = y / 2
plt.plot(x,y,'k')
plt.xticks(np.linspace(-2,2,11),labels=np.round(np.linspace(-2,2,11),2),fontsize=16,**csfont);
plt.yticks(np.linspace(-1,1,11),labels=np.round(np.linspace(-1,1,11),2),fontsize=16,**csfont);
plt.xlabel('input',fontsize=24,**csfont,labelpad=20);
plt.ylabel('output',fontsize=24,**csfont);
plt.text(-3,1.2,'(E)',fontsize=32,**csfont);
plt.title('leaky ReLU\n',fontsize=24,**csfont);
plt.grid()
plt.savefig('../../assets/matplotlib//022_activation_function_3.png', dpi=300,bbox_inches="tight")
```

### Tanh

```python
plt.figure(figsize=(6,6));
x = np.linspace(-2,2,101)
y = np.tanh(x)
plt.plot(x,y,'k')
plt.xticks(np.linspace(-2,2,11),labels=np.round(np.linspace(-2,2,11),2),fontsize=16,**csfont);
plt.yticks(np.linspace(-1,1,11),labels=np.round(np.linspace(-1,1,11),2),fontsize=16,**csfont);
plt.xlabel('input',fontsize=24,**csfont);
plt.ylabel('output',fontsize=24,**csfont);
plt.text(-3,1.2,'(C)',fontsize=32,**csfont);
plt.title('Tanh\n',fontsize=24,**csfont);
plt.grid()
plt.savefig('../../assets/matplotlib//022_activation_function_4.png', dpi=300,bbox_inches="tight")
```

### Sigmoid

```python
plt.figure(figsize=(6,6));
x = np.linspace(-4,4,101)
y = 1 / (1 + np.exp(-x))
plt.plot(x,y,'k')
plt.xticks(np.linspace(-4,4,11),labels=np.round(np.linspace(-4,4,11),2),fontsize=16,**csfont);
plt.yticks(np.linspace(0,1,11),labels=np.round(np.linspace(0,1,11),2),fontsize=16,**csfont);
plt.xlabel('input',fontsize=24,**csfont);
plt.ylabel('output',fontsize=24,**csfont);
plt.text(-5,1.1,'(B)',fontsize=32,**csfont);
plt.title('Sigmoid\n',fontsize=24,**csfont);
plt.grid()
plt.savefig('../../assets/matplotlib//022_activation_function_5.png', dpi=300,bbox_inches="tight")
```
