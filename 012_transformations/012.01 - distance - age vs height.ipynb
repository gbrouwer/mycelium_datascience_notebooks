### Import

```python
import os
import numpy as np
import sys
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from scipy.stats import multivariate_normal as mnorm
from mpl_toolkits import mplot3d
from matplotlib import font_manager as fm, rcParams
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Plot

```python
height_mean = 5
height_var = 1.5**2
height_std = np.sqrt(height_var)
height_domain = [0,10]
age_mean = 50
age_var = 15**2
age_std = np.sqrt(age_var)
age_domain = [0,100]
rv_age_height = mnorm([age_mean,height_mean], [[age_var,0.0], [0.0, height_var]])
```

### Create Levels and Colors

```python
std_range = np.linspace(-2,0,11)
levels_age_height = []
for std_val in std_range:
    prob_std = rv_age_height.pdf(
        [age_mean + (age_std*std_val),
         height_mean + (height_std*std_val)])
    levels_age_height.append(prob_std)
colors_age_height = np.zeros((len(levels_age_height),3)) + 0.6
colors_age_height[5,:] = [0,0,0]
colors_age_height[0,:] = [0,0,0]
```

### Create Contour Plot

```python
resolution = 111
[X,Y] = np.meshgrid(
    np.linspace(
        age_domain[0],
        age_domain[1],
        resolution),
    np.linspace(
        height_domain[0],
        height_domain[1],
        resolution))
XY = np.vstack((X.flatten(),Y.flatten())).T
P = rv_age_height.pdf(XY)
P = np.reshape(P,(resolution,resolution))
N = rv_age_height.rvs(200)
C = np.zeros((200,3))
C[:,0] = 0.3
C[:,1] = 0.6
C[:,2] = 0.9
```

### Plot

```python
plt.figure(figsize=(8,8.5))
plt.contour(P,levels_age_height,extent=(age_domain[0],age_domain[1],height_domain[0],height_domain[1]),
            colors=colors_age_height,
            linestyles='dashed',
            zorder=-100);
plt.scatter(age_mean,height_mean,c='k',marker='+',s=100)
plt.scatter(N[:,0],N[:,1],c=C,s=100)
plt.grid();
plt.ylabel('height (feet)',fontsize=24,**csfont);
plt.xlabel('age (years)',fontsize=24,**csfont);
plt.xticks(fontsize=18,**csfont);
plt.yticks(fontsize=18,**csfont);
plt.axis((age_domain[0],age_domain[1],height_domain[0],height_domain[1]))

plt.scatter(age_mean-30,height_mean,s=300,zorder=1000,linewidth=3,edgecolor='k',facecolor='w')
plt.arrow(age_mean,height_mean,-28,0,
          length_includes_head = True,head_width = 0.5,head_length = 3.5,
          shape = 'full',head_starts_at_zero = False,
          facecolor = 'k',edgecolor = 'k',linewidth=2)
txt = plt.text(age_mean-30-10,height_mean-0.25,'B',c=(0,0,0),fontsize=48,**csfont);

plt.scatter(age_mean,height_mean,s=300,zorder=1000,linewidth=3,edgecolor='k',facecolor='w')
txt = plt.text(age_mean+5,height_mean-0.25,'A',c=(0,0,0),fontsize=48,**csfont);

plt.scatter(age_mean,height_mean+3,s=300,zorder=1000,linewidth=3,edgecolor='k',facecolor='w')
plt.arrow(age_mean,height_mean,0,2.25,
          length_includes_head = False,head_width = 4,head_length = 0.5,
          shape = 'full',head_starts_at_zero = False,
          facecolor = 'k',edgecolor = 'k',linewidth=2)
txt = plt.text(age_mean+5,height_mean+2.75,'C',c=(0,0,0),fontsize=48,**csfont);
plt.savefig('../../assets/figures/012/012_distribution_01.png',dpi=300,bbox_inches="tight");
```
