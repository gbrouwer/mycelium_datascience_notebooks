### Import

```python
import os
import numpy as np
import sys
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from scipy.stats import multivariate_normal as mnorm
from mpl_toolkits import mplot3d
from matplotlib import font_manager as fm, rcParams
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Create a asymmetrical distribution of hours per month exercised and BMI

```python
exercise_mean = 50
exercise_var = 16**2
exercise_std = np.sqrt(exercise_var)
exercise_domain = [0,exercise_mean+exercise_std*4]
bmi_mean = 24
bmi_var = 4**2
bmi_std = np.sqrt(bmi_var)
bmi_domain = [0,bmi_mean+bmi_var*4]
rv_exercise_bmi = mnorm([exercise_mean,bmi_mean], [[exercise_var,-7], [-7, bmi_var]])
```

### Create A substantial amount of data

```python
X = rv_exercise_bmi.rvs(1000000)
```

### Plot Unnormalized Data

```python
exercise_hist = np.histogram(X[:,0],np.linspace(exercise_domain[0],exercise_domain[1],101))
exercise_rescale = exercise_hist[0] / np.sum(exercise_hist[0])
bmi_hist = np.histogram(X[:,1],np.linspace(bmi_domain[0],exercise_domain[1],101))
bmi_rescale = bmi_hist[0] / np.sum(bmi_hist[0])

plt.figure(figsize=(8,8))
plt.plot(exercise_hist[1][1:],exercise_rescale,'r',linewidth=2)
plt.plot(bmi_hist[1][1:],bmi_rescale,'g',linewidth=2)
plt.xlabel('exercise / BMI',fontsize=24,**csfont);
plt.ylabel('Probabiity',fontsize=24,**csfont);
plt.xticks(fontsize=18,**csfont);
plt.yticks(fontsize=18,**csfont);
plt.grid()
plt.savefig('../../assets/figures/012/012_normalization_before.png',dpi=300,bbox_inches="tight");
```

### Normalize

```python
Xnorm = X.copy()
Xnorm[:,0] = Xnorm[:,0] - np.mean(Xnorm[:,0])
Xnorm[:,1] = Xnorm[:,1] - np.mean(Xnorm[:,1])
Xnorm[:,0] = Xnorm[:,0] / np.std(Xnorm[:,0])
Xnorm[:,1] = Xnorm[:,1] / np.std(Xnorm[:,1])
```

### Plot Nornalized Data

```python
exercise_hist = np.histogram(Xnorm[:,0],np.linspace(-3,3,101))
exercise_rescale = exercise_hist[0] / np.sum(exercise_hist[0])
bmi_hist = np.histogram(Xnorm[:,1],np.linspace(-3,3,101))
bmi_rescale = bmi_hist[0] / np.sum(bmi_hist[0])
plt.figure(figsize=(8,8))
plt.plot(exercise_hist[1][1:],exercise_rescale,'r',linewidth=2)
plt.plot(bmi_hist[1][1:],bmi_rescale,'g',linewidth=2)
plt.xlabel('exercise / BMI',fontsize=24,**csfont);
plt.ylabel('Probabiity',fontsize=24,**csfont);
plt.xticks(fontsize=18,**csfont);
plt.yticks(fontsize=18,**csfont);
plt.grid()
plt.savefig('../../assets/figures/012/012_normalization_after.png',dpi=300,bbox_inches="tight");
```
