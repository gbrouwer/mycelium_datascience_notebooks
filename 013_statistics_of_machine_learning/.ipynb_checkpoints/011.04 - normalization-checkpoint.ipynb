### Import

```python
import os
import numpy as np
import sys
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from scipy.stats import multivariate_normal as mnorm
from mpl_toolkits import mplot3d
from matplotlib import font_manager as fm, rcParams
np.random.seed(182)
```

### Create Data

```python
x = np.random.random((100))
y = 2*x + 1 + np.random.normal(0,1,100)*0.25
```

### Fitted Lines

```python
xval1 = np.array([0,1])
yval1 = 2*xval1

xval2 = np.array([0,1])
yval2 = 2*xval2 + 1
```

### Plot

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y,c='k',s=50)
plt.axis((0,1,0,4))
plt.grid()
plt.xlabel('x',fontsize=18)
plt.ylabel('y',fontsize=18)
plt.plot(xval1,yval1,'r--')
plt.plot(xval1,yval2,'g--')
plt.text(0.6,0.5,'y = 2x',fontsize=18,c='r')
plt.arrow(0.35, 2.8,0.15, -0.80,color='g',width=0.01,head_width=0.05,length_includes_head=True)
plt.text(0.2,3.0,'y = 2x + 1',fontsize=18,c='g')
plt.arrow(0.7, 0.75,-0.08, 0.50,color='r',width=0.01,head_width=0.05,length_includes_head=True)
```
