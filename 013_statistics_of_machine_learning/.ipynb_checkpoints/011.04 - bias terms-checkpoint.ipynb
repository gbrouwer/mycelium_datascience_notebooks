### Import

```python
import os
import numpy as np
import sys
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from scipy.stats import multivariate_normal as mnorm
from mpl_toolkits import mplot3d
from matplotlib import font_manager as fm, rcParams
np.random.seed(182)
```

### Original Data

```python
X = np.ones((5,3))
X[0,0] = -5
X[1,0] = -5
X[2,0] = 5
X[3,0] = 5
X[4,0] = 0
X[0,1] = -5
X[1,1] = 5
X[2,1] = 5
X[3,1] = -5
X[4,1] = 0
```

### Rotation Matrix

```python
R = np.eye(3)
degree_angle = 30
rad_angle = degree_angle / 180 * np.pi
R[:2,:2] = np.array([[np.cos(rad_angle),-np.sin(rad_angle)],[np.sin(rad_angle),np.cos(rad_angle)]])
```

### Translation Matrix

```python
T = np.eye(3)
T[2,0] = 5
T[2,1] = -10
```

### Scaling Matrix

```python
S = np.eye(3)
S[0,0] = 1.50
S[1,1] = 1.25
```

### Shear Matrix

```python
H = np.eye(3)
H[0,1] = 0.15
H[1,0] = 0.15
```

### Apply only one

```python
XR = np.dot(X,R)
XS = np.dot(X,S)
XT = np.dot(X,T)
XH = np.dot(X,H)
print(H)
print(XH)
plt.figure(figsize=(6,6));
plt.scatter(X[:,0],X[:,1],s=100,zorder=100,color='k');
plt.grid()
plt.axis((-20,20,-20,20))
plt.savefig('../../assets/figures/11_trans_orig.png');

plt.figure(figsize=(6,6));
plt.scatter(XR[:,0],XR[:,1],s=100,zorder=100,color='k');
plt.grid()
plt.axis((-20,20,-20,20))
plt.savefig('../../assets/figures/11_trans_rot.png');

plt.figure(figsize=(6,6));
plt.scatter(XS[:,0],XS[:,1],s=100,zorder=100,color='k');
plt.grid()
plt.axis((-20,20,-20,20))
plt.savefig('../../assets/figures/11_trans_scale.png');

plt.figure(figsize=(6,6));
plt.scatter(XH[:,0],XH[:,1],s=100,zorder=100,color='k');
plt.grid()
plt.axis((-20,20,-20,20))
plt.savefig('../../assets/figures/11_trans_shear.png');

plt.figure(figsize=(6,6));
plt.scatter(XT[:,0],XT[:,1],s=100,zorder=100,color='k');
plt.grid()
plt.axis((-20,20,-20,20))
plt.savefig('../../assets/figures/11_trans_translation.png');
```

### Apply as Sequence

```python
XR = np.dot(X,R)
XRS = np.dot(XR,S)
XRSH = np.dot(XRS,S)
XRSHT = np.dot(XRSH,T)
M = np.dot(np.dot(np.dot(R,S),H),T)
XM = np.dot(X,M)
print(XM)

plt.figure(figsize=(6,6));
plt.scatter(XM[:,0],XM[:,1],s=100,zorder=100,color='k');
plt.grid()
plt.axis((-20,20,-20,20))
plt.savefig('../../assets/figures/11_trans_combined.png');
```

### Translate Animation

```python
plt.figure(figsize=(6,6));
x_increment = 5/50
y_increment = -10/50
XT = X.copy()
for i in range(50):
    XT[:,0] = x_increment + XT[:,0]
    XT[:,1] = y_increment + XT[:,1]    
    plt.scatter(XT[:,0],XT[:,1],s=100,zorder=100,color='k');
    plt.axis((-20,20,-20,20))
    plt.grid();
    plt.savefig('../../assets/gifs/011/translation/' + str(i).zfill(3) + '.png');
    plt.clf();
```

### Rotation Animation

```python
degree_angle = 30/50
rad_angle = degree_angle / 180 * np.pi
R = np.eye(3)
R[:2,:2] = np.array([[np.cos(rad_angle),-np.sin(rad_angle)],[np.sin(rad_angle),np.cos(rad_angle)]])
XR = X.copy()
plt.figure(figsize=(6,6));
for i in range(50):
    XR = np.dot(XR,R)
    plt.scatter(XR[:,0],XR[:,1],s=100,zorder=100,color='k');
    plt.axis((-20,20,-20,20));
    plt.grid();
    plt.savefig('../../assets/gifs/011/rotation/' + str(i).zfill(3) + '.png');
    plt.clf();
```

### Scale Animation

```python
plt.figure(figsize=(6,6));
XS = X.copy()
for i in range(50):
    ratio_x = 1 + (i/50 * 0.5)
    ratio_y = 1 + (i/50 * 0.25)
    XS[:,0] = X[:,0] * ratio_x
    XS[:,1] = X[:,1] * ratio_y
    plt.scatter(XS[:,0],XS[:,1],s=100,zorder=100,color='k');
    plt.axis((-20,20,-20,20));
    plt.grid();
    plt.savefig('../../assets/gifs/011/scaling/' + str(i).zfill(3) + '.png');
    plt.clf();
```

### Share

```python
plt.figure(figsize=(6,6));
XH = X.copy()
for i in range(50):
    sx = i/50*0.15
    sy = i/50*0.15
    R = np.eye(3)
    R[0,1] = sx
    R[1,0] = sy
    XH = np.dot(X,R)
    plt.scatter(XH[:,0],XH[:,1],s=100,zorder=100,color='k');
    plt.axis((-20,20,-20,20));
    plt.grid();
    plt.savefig('../../assets/gifs/011/shear/' + str(i).zfill(3) + '.png');
    plt.clf();
```

### Combined

```python
M = [[1.20528811,-0.43014428,0.0000],
     [0.91237976,1.19503175,0.0],
     [5.,-10.,1.]]
M = np.array(M)
plt.figure(figsize=(6,6));
XM = X.copy()
for i in range(50):
    ratio = i/50
    Mthis = np.eye(3)*(1-ratio) + M*ratio
    XM = np.dot(X,Mthis)
    plt.scatter(XM[:,0],XM[:,1],s=100,zorder=100,color='k');
    plt.axis((-20,20,-20,20));
    plt.grid();
    plt.savefig('../../assets/gifs/011/transformation/' + str(i).zfill(3) + '.png');
    plt.clf();
```

### Show Matrix Dot Product Identity

```python
I = np.eye(3)
XR = X.copy()
plt.figure(figsize=(12,6));
count = 0
for n in range(5):
    for m in range(3):
        count = count + 1
        row = X[n,:]
        column = I[:,m]
        result = XR[n,m]
        compstr1 = str(row[0]) + "*" + str(column[0])
        compstr2 = str(row[1]) + "*" +  str(column[1])
        compstr3 = str(row[2]) + "*" +  str(column[2])
        compstr = compstr1 + " + " + compstr2 + " + " + compstr3 + ' = ' + str(result)

        for i in range(X.shape[0]):
            for j in range(X.shape[1]):
                if (i==n):
                    plt.text(j*3+0.5,-i,str(X[i,j]),fontsize=24,c='k',fontweight='bold')
                else:
                    plt.text(j*3+0.5,-i,str(X[i,j]),fontsize=24,c='k')

        for i in range(I.shape[0]):
            for j in range(I.shape[1]):
                if (j==m):
                    plt.text(12.5+(j*3),-i-1,str(I[i,j]),fontsize=24,c='k',fontweight='bold')
                else:
                    plt.text(12.5+(j*3),-i-1,str(I[i,j]),fontsize=24,c='k')

        for i in range(XR.shape[0]):
            for j in range(XR.shape[1]):
                if (j==m and i == n):
                    plt.text(24.5+(j*3),-i,str(XR[i,j]),fontsize=24,c='k',fontweight='bold')
                else:
                    plt.text(24.5+(j*3),-i,str(XR[i,j]),fontsize=24,c='k')

        plt.plot([-0.5,-0.5],[0.75,-4.25],'k')
        plt.plot([-0.5,0.5],[0.75,0.75],'k')
        plt.plot([-0.5,0.5],[-4.25,-4.25],'k')
        plt.plot([9.5,9.5],[0.75,-4.25],'k')
        plt.plot([9.5,8.5],[0.75,0.75],'k')
        plt.plot([9.5,8.5],[-4.25,-4.25],'k')
        plt.plot([23.5,23.5],[0.75,-4.25],'k')
        plt.plot([23.5,24.5],[0.75,0.75],'k')
        plt.plot([23.5,24.5],[-4.25,-4.25],'k')
        plt.plot([33.5,33.5],[0.75,-4.25],'k')
        plt.plot([33.5,32.5],[0.75,0.75],'k')
        plt.plot([33.5,32.5],[-4.25,-4.25],'k')
        plt.plot([11.5,11.5],[-0.25,-3.25],'k')
        plt.plot([11.5,12.5],[-0.25,-0.25],'k')
        plt.plot([11.5,12.5],[-3.25,-3.25],'k')
        plt.plot([21.5,21.5],[-0.25,-3.25],'k')
        plt.plot([21.5,20.5],[-0.25,-0.25],'k')
        plt.plot([21.5,20.5],[-3.25,-3.25],'k')
        plt.scatter(3.5+(j*3.5),-i+2.15,s=50,c='k')
        plt.text(15+(j*3.5),-i+2,"=",fontsize=24)
        plt.text(5,-i-2,compstr,fontsize=24,fontweight='bold')
        plt.axis((-1,34,-7,2)) 
        plt.axis('off');
        plt.savefig('../../assets/gifs/011/dotproduct_identify/' + str(count).zfill(3) + '.png');
        plt.clf();        
```
