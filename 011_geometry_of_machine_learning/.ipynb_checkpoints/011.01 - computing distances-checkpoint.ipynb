### Import

```python
import os
import numpy as np
import sys
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from scipy.stats import multivariate_normal as mnorm
from mpl_toolkits import mplot3d
np.random.seed(182)
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Single Distribution showing the problem of scaling
Frequently, our regressors will represent hugely different quantities, measurements, scales, domains and ranges. Normalization of individual regressors by centering them on the mean and adjusting their variances to be standardized (z-scoring) can mitigate our machine learning models from running away with the regressor with the largest values and range. 

```python
mu = [5.5,50]
sigma = [[0.5,0],[0,50]]
rv = mnorm(mu,sigma)
x = np.linspace(2,8,101)
y = np.linspace(25,75,101)
X,Y = np.meshgrid(x, y)
XX = np.array([X.ravel(), Y.ravel()]).T
Z = rv.pdf(XX).reshape(X.shape)
P = rv.rvs(250)
plt.figure(figsize=(6,6));
plt.contour(X,Y,Z,colors='gray')
plt.scatter(P[:,0],P[:,1],zorder=100,color='r');
plt.axis((2,8,25,75));
plt.scatter(5.5,50,color='k',s=100,zorder=101);
plt.plot([4.5,5.5],[50,50],c='k',zorder=101)
plt.plot([5.5,5.5],[50,60],c='k',zorder=101)
plt.scatter(4.5,50,color='k',s=100,zorder=101);
plt.scatter(5.5,60,color='k',s=100,zorder=101);
plt.xlabel('height (feet)',fontsize=18,**csfont);
plt.ylabel('age (years)',fontsize=18,**csfont);
plt.text(4.25,44,"B",fontsize=32,zorder=102,**csfont);
plt.text(5.25,44,"A",fontsize=32,zorder=102,**csfont);
plt.text(5.25,63,"C",fontsize=32,zorder=102,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.grid()
plt.savefig('../../assets/figures/011/distance_01.png',bbox_inches="tight")
```

### Single Distribution showing the problem of unequal variance
Even when our regressors do represent the same quantity or measurement, there is still the possibility that one shows a much larger variance. Again, normalization can greatly reduce the ill-effects of such discrepancies.

```python
mu = [15,15]
sigma = [[10,0],[0,40]] 
rv = mnorm(mu,sigma)
x = np.linspace(0,30,101)
y = np.linspace(0,30,101)
X,Y = np.meshgrid(x, y)
XX = np.array([X.ravel(), Y.ravel()]).T
Z = rv.pdf(XX).reshape(X.shape)
P = rv.rvs(250)

plt.figure(figsize=(6,6));
plt.contour(X,Y,Z,colors='gray')
plt.scatter(P[:,0],P[:,1],zorder=100,color='r');
plt.scatter(15,20,color='k',s=100,zorder=101);
plt.scatter(15,15,color='k',s=100,zorder=101);
plt.scatter(10,15,color='k',s=100,zorder=101);
plt.plot([10,15],[15,15],c='k',zorder=101)
plt.plot([15,15],[15,20],c='k',zorder=101)
plt.axis((0,30,0,30));
plt.xlabel('years of education',fontsize=18,**csfont);
plt.ylabel('years of mariage',fontsize=18,**csfont);
plt.text(9.6,11.5,"B",fontsize=32,zorder=102,**csfont);
plt.text(14.6,11.5,"A",fontsize=32,zorder=102,**csfont);
plt.text(14.6,21,"C",fontsize=32,zorder=102,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.grid()
plt.savefig('../../assets/figures/011/distance_02.png',bbox_inches="tight")
```

### Single Distribution showing the problem of covariance
Covariance is hard to deal with. Covariance between regresors is never desirable. And measures of strict euclidean distance are misleading. 

```python
mu = [15,15]
sigma = [[10,-10],[-10,12]] 
rv = mnorm(mu,sigma)
x = np.linspace(0,30,101)
y = np.linspace(0,30,101)
X,Y = np.meshgrid(x, y)
XX = np.array([X.ravel(), Y.ravel()]).T
Z = rv.pdf(XX).reshape(X.shape)
P = rv.rvs(250)

plt.figure(figsize=(6,6));
plt.contour(X,Y,Z,colors='gray')
plt.scatter(P[:,0],P[:,1],zorder=100,color='r');
plt.scatter(20,20,color='k',s=100,zorder=101);
plt.scatter(15,15,color='k',s=100,zorder=101);
plt.scatter(10,20,color='k',s=100,zorder=101);
plt.plot([10,15],[20,15],c='k',zorder=101)
plt.plot([15,20],[15,20],c='k',zorder=101)
plt.xlabel('days of month exercised',fontsize=18,**csfont);
plt.ylabel('BMI',fontsize=18,**csfont);
plt.text(7.6,16.5,"B",fontsize=32,zorder=102,**csfont);
plt.text(14.6,11.5,"A",fontsize=32,zorder=102,**csfont);
plt.text(19.6,21,"C",fontsize=32,zorder=102,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.grid()
plt.savefig('../../assets/figures/011/distance_03.png',bbox_inches="tight")
```
