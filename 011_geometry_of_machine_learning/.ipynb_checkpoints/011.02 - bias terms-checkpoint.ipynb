### Import

```python
import os
import numpy as np
import sys
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from scipy.stats import multivariate_normal as mnorm
from mpl_toolkits import mplot3d
from matplotlib import font_manager as fm, rcParams
np.random.seed(182)
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Create Data

```python
x = np.random.random((100))
y = 2*x + 1 + np.random.normal(0,1,100)*0.25
```

### Fitted Lines

```python
xval1 = np.array([0,1])
yval1 = 2*xval1

xval2 = np.array([0,1])
yval2 = 2*xval2 + 1
```

### Plot

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y,c='k',s=50)
plt.axis((0,1,0,4))
plt.grid()
plt.xlabel('x',fontsize=24)
plt.ylabel('y',fontsize=24)
plt.plot(xval1,yval1,'r',linewidth=2)
plt.plot(xval1,yval2,'g',linewidth=2)
plt.text(0.6,0.9,'y = 2x',fontweight=600,fontsize=24,c='r',**csfont);
plt.text(0.2,2.6,'y = 2x + 1',fontweight=600,fontsize=24,c='g',**csfont);
plt.xlabel('x',fontsize=22,**csfont);
plt.ylabel('y',fontsize=22,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.savefig('../../assets/figures/011/bias.png',bbox_inches="tight")
```
