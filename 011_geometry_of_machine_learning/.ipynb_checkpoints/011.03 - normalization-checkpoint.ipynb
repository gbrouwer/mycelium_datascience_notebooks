### Import

```python
import os
import numpy as np
import sys
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from scipy.stats import multivariate_normal as mnorm
from mpl_toolkits import mplot3d
from matplotlib import font_manager as fm, rcParams
np.random.seed(182)
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Create Data

```python
O = np.random.normal(100,25,10000)
```

### Histogram

```python
plt.figure(figsize=(6,6));
h1 = np.histogram(O,np.linspace(0,200,51));
plt.bar(h1[1][1:],h1[0]/np.sum(h1[0]),4,edgecolor='k',facecolor='w');
plt.xlabel('x',fontsize=24,**csfont);
plt.ylabel('p(x)',fontsize=24,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.savefig('../../assets/figures/011/normalization_01.png')
```

```python
O1 = O - np.mean(O)
O2 = O1 / np.std(O1)
```

### Plot Again

```python
plt.figure(figsize=(6,6));
h1 = np.histogram(O2,np.linspace(-5,5,51));
plt.bar(h1[1][1:],h1[0]/np.sum(h1[0]),0.2,edgecolor='k',facecolor='w');
plt.xlabel('x',fontsize=24,**csfont);
plt.ylabel('p(x)',fontsize=24,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.savefig('../../assets/figures/011/normalization_02.png')
```
