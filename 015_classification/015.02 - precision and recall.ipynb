### Fonts

### Plot

```python
plt.figure(figsize=(14,10))
plt.plot([0,0],[-10,10],'k')
plt.plot([10,10],[-10,10],'k')
plt.plot([-10,-10],[-10,10],'k')
plt.plot([-10,10],[-10,-10],'k')
plt.plot([-10,10],[0,0],'k')
plt.plot([-10,10],[10,10],'k')
plt.text(-8,5,'True Positive',fontsize=22,**csfont);
plt.text(-8,-5,'False Negative',fontsize=22,**csfont);
plt.text(2,5,'False Positive',fontsize=22,**csfont);
plt.text(2,-5,'True Negative',fontsize=22,**csfont);

plt.text(-12,5,'Yes',fontsize=22,**csfont);
plt.text(-12,-5,'No',fontsize=22,**csfont);
plt.text(-6,11,'Yes',fontsize=22,**csfont);
plt.text(4,11,'No',fontsize=22,**csfont);
plt.text(-13.5,-4,'Predicted outcome',fontsize=22,**csfont,rotation=90);
plt.text(-4,12.5,'Actual outcome',fontsize=22,**csfont)
plt.text(18.5,6,'TP',fontsize=22,**csfont);plt.plot([-10,10],[10,10],'k')
plt.text(12,5,'Precision =',fontsize=22,**csfont);
plt.plot([17,21],[5.3,5.3],'k')
plt.text(17.3,4,'TP + FP',fontsize=22,**csfont);
plt.text(-3.5,-12,'TP',fontsize=22,**csfont);plt.plot([-10,10],[10,10],'k')
plt.text(-5,-14,'TP + FN',fontsize=22,**csfont);
plt.text(-9,-13,'Recall =',fontsize=22,**csfont);
plt.plot([-5.2,-1.2],[-12.65,-12.65],'k')
plt.axis('off');
plt.savefig('../../assets/matplotlib/015_precision_recall.png',dpi=300,bbox_inches="tight")
```
