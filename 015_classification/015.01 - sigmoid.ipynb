### Import

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Sigmoid function

```python
def sigmoid(x):
    return 1  / (1 + np.exp(-x))
```

### Plot Sigmoid

```python
x = np.linspace(-10,10,128)
s = sigmoid(x)
plt.figure(figsize=(8,6))
plt.plot(x,s,linewidth=4,c='k')
plt.xlabel('x',fontsize=22,**csfont);
plt.ylabel('sigmoid(x)',fontsize=22,**csfont);
plt.xticks(fontsize=18,**csfont);
plt.yticks(fontsize=18,**csfont);
plt.grid()
plt.savefig('../../assets/matplotlib/015_sigmoid.png',dpi=300,bbox_inches="tight")
```
