### Import

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Loss function

```python
def compute_loss(y,p):
    loss = -y*np.log(p) + (1 - y) * np.log(1-p)
    return loss
```

### Loss Landscape

```python
x = np.linspace(0.00001,1-0.00001,101)
pos = np.abs(compute_loss(1,x))
neg = np.abs(compute_loss(0,x))
```

### Plot

```python
plt.figure(figsize=(8,6))
plt.plot(x,pos,'k',linewidth=3);
plt.plot(x,neg,c=(0.75,0.75,0.75),linewidth=3);
plt.ylabel('loss',fontsize=18);
plt.grid()
plt.xlabel('prob(y|x)',fontsize=22,**csfont);
plt.ylabel('loss',fontsize=22,**csfont);
plt.xticks(fontsize=18,**csfont);
plt.yticks(fontsize=18,**csfont);
plt.savefig('../../assets/matplotlib/015_log_loss.png',dpi=300,bbox_inches="tight")
```
