### Import

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

### Sigmoid function

```python
def sigmoid(x):
    return 1  / (1 + np.exp(-x))
```

### Plot Sigmoid

```python
x = np.linspace(-10,10,128)
s = sigmoid(x)
plt.figure(figsize=(6,6))
for i in range(len(x)-1):
    x1 = x[i]
    x2 = x[i+1]
    s1 = s[i]
    s2 = s[i+1]
    plt.plot([x1,x2],[s1,s2],c=(i/128,1-(i/256),0.75),linewidth=4)
plt.xlabel('x',fontsize=18)
plt.ylabel('sigmoid(x)',fontsize=18)
plt.grid()
plt.savefig('../../assets/figures/14_sigmoid.png')
```
