# Analyzing Mozart
Here, we look at the frequencies of Mozart's Clarinet Concerto in A Major (K-622). Can we see the signature of this key in the FFT analysis?

```python
import os
import numpy as np
import sys
import pandas as pd
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa.display
import librosa
from scipy import fft
from scipy import ifft
from scipy.signal import find_peaks
from matplotlib import cm
from tqdm import tqdm
from scipy.fft import irfft, rfft, rfftfreq
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Load Waveform (first 30 seconds in stereo)

```python
signal, sr = librosa.load('../../data/audio/mozart_clarinet_concerto_a_major_k622.wav',mono=False,duration=30.0)
left = signal[0,:]
right = signal[1,:]
ipd.Audio(left, rate=sr)
```

### Plot the timeseries

```python
plt.figure(figsize=(14, 5))
plt.plot(left,c='k')
plt.xlabel('Time (ms)',fontsize=24,**csfont);
plt.ylabel('Voltage',fontsize=24,**csfont);
xtick = np.linspace(0,len(left),11)
ytick = np.linspace(-0.15,0.15,11)
plt.xticks(xtick,labels=np.round(xtick/sr,2),fontsize=18,**csfont);
plt.yticks(ytick,labels=np.round(ytick,2),fontsize=18,**csfont);
plt.savefig('../../assets/matplotlib/018_mozart_waveform.png',dpi=300,bbox_inches="tight")
```

### We use fft to move into the frequency domain
We take the 30seconds of audio and transform it to the frequency domain. There, we find the peak frequencies.

```python
plt.figure(figsize=(14, 5))
cutoff_freq = 2000
N = len(left)
yf = rfft(left)
xf = rfftfreq(N, 1 / sr)
power = np.abs(yf)
power = power / np.max(power)
cutoff_idx = np.where(xf >= cutoff_freq)[0][0]
xf = xf[:cutoff_idx]
power = power[:cutoff_idx]
peaks = find_peaks(power,distance=1000,threshold=0.01)[0]
peaks_freq = xf[peaks];
fxtick = np.linspace(0,cutoff_freq,11)
fytick = np.linspace(0,1.5,7)
fyticklabels = fytick.copy().astype(str)
fyticklabels[-1] = ''
fyticklabels[-2] = ''
plt.plot(xf, power,c='k')
plt.xlabel('Frequency (Herz)',fontsize=24,**csfont);
plt.ylabel('Power (Arbitrary Units)',fontsize=24,**csfont);
plt.scatter(peaks_freq,power[peaks],c='r',s=50)
for i,peak in enumerate(peaks):
    peak_freq = peaks_freq[i]
    plt.text(peak_freq-25,power[peak]+0.1,str(np.round(peak_freq).astype(int)),rotation=90,fontsize=14,**csfont);
plt.xticks(fxtick,labels=fxtick.astype(int),fontsize=18,**csfont);
plt.yticks(fytick,labels=fyticklabels,fontsize=18,**csfont);
plt.savefig('../../assets/matplotlib/018_mozart_freq_domain.png',dpi=300,bbox_inches="tight")
```

# Spectogram

### Create Spectrogram and plot

```python
hop_length = 512
X = librosa.stft(left)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
ax = librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz', hop_length=hop_length);
ax.set_cmap('viridis')
plt.xlabel('Time (ms)',fontsize=24,**csfont);
plt.ylabel('Frequency',fontsize=24,**csfont);
plt.yticks(np.linspace(0,2000,5),labels=np.linspace(0,2000,5).astype(int),fontsize=18,**csfont);
plt.xticks(np.linspace(0,len(left)/sr,13),fontsize=18,**csfont);
plt.axis((0,30,0,2000));
plt.savefig('../../assets/matplotlib/018_mozart_spectrogram.png',dpi=300,bbox_inches="tight")
```
