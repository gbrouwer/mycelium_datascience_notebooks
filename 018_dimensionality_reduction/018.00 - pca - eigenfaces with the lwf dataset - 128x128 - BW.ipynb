### Run PCA
Run the PCA on the all the faces, stacked together into a 2D Matrix. The rows represent individual pixels, the columns the different faces in the data. The PCA will attempt the to find the lower dimensional space of principal componenets as a linear weighting of the original space of the individual faces. In other words, each principal component will be a weighted mixture of all the faces in the data. Some of the principal components will capture the similarity in a subset of the faces, whereas others will capture similarities and variations that all faces share. The biggest similarity between a large collection of pictures of human faces is what they all have more or less in common: two eyes, a nose, a mouth and and overal shape of the face itself. The first principal componenent will capture this 'mean' face as it is the largest source of variance in the data: the individual pixels differ from the mean pixel intensity across all pixels as a function of their location. 

```python
pca = PCA(n_components=no_components)
E = pca.fit_transform(X)
```

### Plot Explained Variance first
After an initial fast increase in cumulative explained variance by the first couple of pricipal components, the curve flattens, indicating the later principal components have dimishiing returns: they explain less and less remaining variance. 

```python
xticks = np.linspace(1,no_components,5)
xticks = xticks.astype(int)
yticks = np.round(np.linspace(0,1,11),2)
plt.figure(figsize=(6,6))
plt.subplots_adjust(left=0.0,bottom=0.0, right=1.0, top=1.0, wspace=0.1, hspace=0.1)
plt.plot(range(1,no_components+1),np.cumsum(pca.explained_variance_ratio_),'k');
plt.scatter(range(1,no_components+1),np.cumsum(pca.explained_variance_ratio_),c='r',s=100,edgecolor='k')
plt.grid()
plt.axis((0,no_components+1,0,1))
plt.ylabel('Cumulative Explained Variance (%)',fontsize=24,**csfont);
plt.ylabel('Component Number',fontsize=18,**csfont);
plt.xticks(xticks,labels = xticks,fontsize=18,**csfont);
plt.yticks(yticks,labels = yticks,fontsize=18,**csfont);
```

### Create a normalized version of E for plotting purposes so that it we can visualize the PC as images

```python
Eviz = np.zeros((E.shape))
for i in range(E.shape[1]):
    pc = E[:,i]
    maxval = np.max([np.abs(np.min(pc)),np.abs(np.max(pc))])
    pc = (pc / maxval * 0.5) + 0.5
    Eviz[:,i] = pc
```

### Resize Test images make BW if necessary

```python
gijs = np.array(gijs.resize((image_size,image_size)))
john = np.array(john.resize((image_size,image_size)))
saoirse = np.array(saoirse.resize((image_size,image_size)))[:,:,:3]
mbappe = np.array(mbappe.resize((image_size,image_size)))
zendaya = np.array(zendaya.resize((image_size,image_size)))[:,:,:3]
gijs_bw = np.mean(gijs,axis=2)
john_bw = np.mean(john,axis=2)
saoirse_bw = np.mean(saoirse,axis=2)
mbappe_bw = np.mean(mbappe,axis=2)
zendaya_bw = np.mean(zendaya,axis=2)
```

### To Vectors

```python
gijs_vector = np.reshape(gijs_bw,(gijs_bw.shape[0]*gijs_bw.shape[1],1))
john_vector = np.reshape(john_bw,(john_bw.shape[0]*john_bw.shape[1],1))
saoirse_vector = np.reshape(saoirse_bw,(saoirse_bw.shape[0]*saoirse_bw.shape[1],1))
mbappe_vector = np.reshape(mbappe_bw,(mbappe_bw.shape[0]*mbappe_bw.shape[1],1))
zendaya_vector = np.reshape(zendaya_bw,(zendaya_bw.shape[0]*zendaya_bw.shape[1],1))
```

### Compute weights of each face on the n compotnents

```python
W1 = np.dot(np.dot(np.linalg.inv(np.dot(E.T,E)),E.T),gijs_vector)
W2 = np.dot(np.dot(np.linalg.inv(np.dot(E.T,E)),E.T),john_vector)
W3 = np.dot(np.dot(np.linalg.inv(np.dot(E.T,E)),E.T),saoirse_vector)
W4 = np.dot(np.dot(np.linalg.inv(np.dot(E.T,E)),E.T),mbappe_vector)
W5 = np.dot(np.dot(np.linalg.inv(np.dot(E.T,E)),E.T),zendaya_vector)

recon_face1 = np.dot(E,W1)
recon_face1 = np.reshape(recon_face1,(imsize,imsize))
recon_face2 = np.dot(E,W2)
recon_face2 = np.reshape(recon_face2,(imsize,imsize))
recon_face3 = np.dot(E,W3)
recon_face3 = np.reshape(recon_face3,(imsize,imsize))
recon_face4 = np.dot(E,W4)
recon_face4 = np.reshape(recon_face4,(imsize,imsize))
recon_face5 = np.dot(E,W5)
recon_face5 = np.reshape(recon_face5,(imsize,imsize))

recon_face1 = recon_face1 - np.min(recon_face1)
recon_face1 = recon_face1 / np.max(recon_face1)
recon_face2 = recon_face2 - np.min(recon_face2)
recon_face2 = recon_face2 / np.max(recon_face2)
recon_face3 = recon_face3 - np.min(recon_face3)
recon_face3 = recon_face3 / np.max(recon_face3)
recon_face4 = recon_face4 - np.min(recon_face4)
recon_face4 = recon_face4 / np.max(recon_face4)
recon_face5 = recon_face5 - np.min(recon_face5)
recon_face5 = recon_face5 / np.max(recon_face5)
```
