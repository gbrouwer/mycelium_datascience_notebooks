### Import

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
from PIL import Image
from scipy.stats import multivariate_normal as mnorm
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Load Image

```python
im = np.array(Image.open("../../data/dct/butterfly_1200x800.bmp"))
```

### Transform the images to gray scale to makes things a bit easier and clearer

```python
im = np.mean(im,axis=2)
```

### Visualize

```python
plt.figure(figsize=(14,14))
plt.imshow(im,cmap='gray')
plt.axis('off')
plt.savefig('../../assets/matplotlib/018_dct_original.png',dpi=300,bbox_inches="tight")
```

### Discrete Cosine Tranform

```python
def dct2(a):
    return fftpack.dct( scipy.fftpack.dct( a, axis=0, norm='ortho' ), axis=1, norm='ortho' )
```

### Inverse Discrete Cosine Transform

```python
def idct2(a):
    return fftpack.idct( scipy.fftpack.idct( a, axis=0 , norm='ortho'), axis=1 , norm='ortho')
```

### Divide the image in blocks of 8x8 and perform DCT on each block

```python
def blockwise_dct(im):

    imsize = im.shape
    dct_im = np.zeros(imsize)

    for i in range(0,imsize[0],8):
        for j in range(0,imsize[1],8):
            dct_im[i:(i+8),j:(j+8)] = dct2(im[i:(i+8),j:(j+8)])

    #Return
    return dct_im
```

### Divide the image in blocks of 8x8 and perform Inverse DCT on each block

```python
def blockwise_idct(dct):

    dctsize = dct.shape
    im = np.zeros(dctsize)

    for i in range(0,dctsize[0],8):
        for j in range(0,dctsize[1],8):
            im[i:(i+8),j:(j+8)] = idct2(dct[i:(i+8),j:(j+8)])

    #Return
    return im
```

### Perform DCT on the image

```python
dct_im = blockwise_dct(im)
```

### Visualize a block

```python
posx = 500
posy = 500
imblock = im[posx:posx+8,posy:posy+8]
dctblock = dct_im[posx:posx+8,posy:posy+8]
dctblock = np.abs(dctblock)
dctblock = np.log(dctblock)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(imblock,cmap='gray',interpolation='nearest')
plt.axis('off');
plt.subplot(1,2,2)
plt.imshow(dctblock,cmap='gray', extent=[0,np.pi,np.pi,0])
plt.axis('off');
plt.savefig('../../assets/matplotlib/018_dct_pixel_and_dct_block.png',dpi=300,bbox_inches="tight")
```

### Display Entire Image as DCT block


```python
plt.figure(figsize=(10,6));
plt.imshow(dct_im,cmap='gray',vmax = np.max(dct_im)*0.01,vmin = 0);
plt.axis('off');
plt.savefig('../../assets/matplotlib/018_dct_image.png',dpi=300,bbox_inches="tight")
```

### Threshold the DCT components and remove the ones that do not make the cut by setting them to zero

```python
thresh = 0.0001
imsize = im.shape

dct_im_thresh = dct_im * (abs(dct_im) > (thresh*np.max(dct_im)))
percent_nonzeros = np.sum( dct_im_thresh != 0.0 ) / (imsize[0]*imsize[1]*1.0)
im_recon = blockwise_idct(dct_im_thresh)

plt.figure(figsize=(12,12))
plt.subplot(2,1,1)
plt.imshow(im,cmap='gray')
plt.axis('off')
plt.title('Low Threshold\n' + str(np.round(percent_nonzeros*100.0,2)) + ' % coefficients retained',**csfont,fontsize=24,pad=20)
plt.subplot(2,1,2)
plt.imshow(im_recon,cmap='gray')
plt.axis('off');
plt.subplots_adjust(bottom=0.1, left=0.1, right=0.9, top=0.9)

print("Retaining  %f%% of the DCT coefficients" % (percent_nonzeros*100.0))
plt.savefig('../../assets/matplotlib/018_dct_low_threshold.png',dpi=300,bbox_inches="tight")
```

### Higher Threshold

```python
thresh = 0.001
dct_im_thresh = dct_im * (abs(dct_im) > (thresh*np.max(dct_im)))
percent_nonzeros = np.sum( dct_im_thresh != 0.0 ) / (imsize[0]*imsize[1]*1.0)
im_recon = blockwise_idct(dct_im_thresh)
print("Retaining  %f%% of the DCT coefficients" % (percent_nonzeros*100.0))

plt.figure(figsize=(12,12))
plt.subplot(2,1,1)
plt.imshow(im,cmap='gray')
plt.title('Medium Threshold\n' + str(np.round(percent_nonzeros*100.0,2)) + ' % coefficients retained',
          **csfont,fontsize=24,pad=20)
plt.axis('off')
plt.subplot(2,1,2)
plt.imshow(im_recon,cmap='gray')
plt.axis('off');
plt.subplots_adjust(bottom=0.1, left=0.1, right=0.9, top=0.9)
plt.savefig('../../assets/matplotlib/018_dct_medium_threshold.png',dpi=300,bbox_inches="tight")
```
