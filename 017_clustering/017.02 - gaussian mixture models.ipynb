### Create Clusters and sample them to generate data

```python
base_mu1 = np.array([+1.5, -1.5])
base_mu2 = np.array([-1.5, +1.5])
base_sigma1 = np.array([[2.0, 0.3], [0.3, 0.5]]) * 0.5
base_sigma2 = np.array([[1.0, 0.3], [0.3, 1.0]]) * 0.5
rv1 = mnormal(base_mu1,base_sigma1)
rv2 = mnormal(base_mu2,base_sigma2)
x1 = rv1.rvs(no_points)
x2 = rv2.rvs(no_points)
X = np.vstack((x1,x2))
```

### Init Algorithm

```python
def initialize(k):

    #Init
    clusters = []
    for idx in range(k):
        cluster = {}
        cluster['pi'] = 1 / k
        cluster['mu'] = (np.random.random((2)) - 0.5) / 10.0
        cluster['sigma'] = np.eye(2)
        clusters.append(cluster)
        
    #Return
    return clusters
```

### Expectation

```python
def expectation(X,clusters):
    
    totals = np.zeros((X.shape[0]))
    for c,cluster in enumerate(clusters):
        pi = cluster['pi']
        mu = cluster['mu']
        sigma = cluster['sigma']
        rv = mnormal(mu,sigma)
        gamma = rv.pdf(X)
        cluster['gamma'] = gamma
        totals = totals + gamma
        clusters[c] = cluster
    for c,cluster in enumerate(clusters):
        cluster['gamma'] = cluster['gamma'] / totals
        clusters[c] = cluster
        
    #Return 
    return X,clusters
```

### Maximization

```python
def maximization(X,clusters):
    
    n = float(X.shape[0])
    for c,cluster in enumerate(clusters):
        gamma = cluster['gamma']
        prev_pi = cluster['pi']
        prev_mu = cluster['mu']
        prev_sigma = cluster['sigma']
        sigma = np.zeros((X.shape[1], X.shape[1]))
        nk = np.sum(gamma, axis=0)
        pi = nk / n
        gamma = np.vstack((gamma,gamma)).T
        mu = np.sum(gamma * X, axis=0) / nk
        for j in range(X.shape[0]):
            diff = (X[j] - mu).reshape(-1, 1)
            sigma += gamma[j] * np.dot(diff, diff.T)
        sigma /= nk
        cluster['pi'] = pi
        cluster['mu'] = mu
        cluster['sigma'] = sigma
        clusters[c] = cluster

    #Return
    return X,clusters
```

### Run

```python
k = 2
history = []
clusters = initialize(k)
clusters_store = copy.deepcopy(clusters)
history.append(clusters_store)
for i in range(30):
    X,clusters = expectation(X,clusters)
    X,clusters = maximization(X,clusters)
    history.append(copy.deepcopy(clusters))
```

### Colors to draw the distribution

### Convert the history 

```python
mu_history = []
sigma_history = []
for item in history:
    mu_list = []
    sigma_list = []
    for element in item:
        mu = element['mu']
        sigma = element['sigma']
        mu_list.append(mu)
        sigma_list.append(sigma)
    mu_history.append(mu_list)
    sigma_history.append(sigma_list)
mu_history = np.array(mu_history)
sigma_history = np.array(sigma_history)
```
