# DBSCAN using Numpy, SciPy and NetworkX
The following is an implementation of DBSCAN using a combination of Numpy, SciPy and NetworkX. The algorithm computes a few graph theoretic measures, which in principle can be implemented without too much effort using NumPy and SciPy. However, using the NetworkX module greatly reduces the number of lines of code we would need, making it easier to read while still clearly showing how the algorithm works. We will discuss graphs in much greater detail in Chapter 19, here we only need a few key ideas to understand what NetworkX computes on behalf of our DBSCAN algorithm.<br>

<br>
A graph is a mathematical model, consisting of nodes and edges. The nodes represent some entity of interest, which can be a human, a word, a location, a variable, or even a mathematical function, to name but a few diverse exmaples. The edges (also called links) indicate some kind of relationship that can exist between two nodes. Two humans can be friends, two words can be synonums, locations can be linked through roads, and varialbes can be linked in how much they are observed to correlate. The links can be undirected, meaning that if ndoe A is connected to B, the reverse is true as well. This is the Facenbook model. If luser A has B as a friend, B has user A as a friend. Twitter is what we call an undirected network: person A can fllow person B, but that does not mean person B follow person A. Edges can also express some continious value. The edge between Boston and New York can be expressed as total distance, or travel time by some mode of transportation. Words can have edges that represent how close they are in their semantic meaning, and edges between different variables can reflect some measure of their correlation strength.<br>
<br>
In other words, the edges in graphs can be conceptual and abstract notions of different kinds of relationships between nodes. However, from a clustering perspective, the edges between nodes (data points) will need to specify some notion of distance between nodes in our data space. In what follows we will use Euclidean distance between 2 data points to represent the strength or weight of the edge connecting the two data point nodes. 

### Import
Standard imports

```python
import os
import sys
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
import networkx as nx
from tqdm import tqdm
from sklearn.datasets import make_moons, make_circles, make_classification
from scipy.stats import multivariate_normal as mnormal
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Parameters
The DBScan algorithm has 2 parameters the user needs to specify before-hand. The parameter eps represents the maximum distance between two points for them to be considered neighbors. Above that distance, nodes cannot be neighbors and as such, cannot be part of the same cluster. The second paramter is minPts, or the minimum of number of neighbors a point must have to be considered a core point. DBSCAN classifies our data points as being either a core point, an edge point or an outlier. Edge points have neighbors that are core points within the allowable distance set by eps, but not enough of them, relative to the minPts parameter. Specified in this way, they will be found around the outer edge of the cluster, hence their name. Outliers are points that are not reachable by any other point, because they have no neighbors within the allowable distance set by eps. 

```python
eps = 0.2
minPts = 5
noise = 0.1
no_samples = 200
random_state = 0
```

### Create Half Moons
We use the half moon dataset generator to create a data set that cannot be accuractely clustered by K-means or GMMs.

```python
X,y = make_moons(noise=noise, random_state=random_state, n_samples=no_samples)
X[:,0] = X[:,0] - np.mean(X[:,0],axis=0)
X[:,1] = X[:,1] - np.mean(X[:,1],axis=0)
X = X * 0.5
X[:,0] = X[:,0] / np.max(X[:,1],axis=0)
X[:,1] = X[:,1] / np.max(X[:,1],axis=0)
```

### Step 1: Find the core points

```python
#Init Corepoints
corepoints = np.zeros((X.shape[0]))

#Compute the distance matrix of our points
D = squareform(pdist(X))

#Threshold D based on eps, creating a binary adjacency matrix, 
#where each i,j element represents whether points i and j are 
#within the eps distance away from eachother. 
A = (D <= eps) * 1

#Create a graph and make sure to exclude self-loops (i -> i)
G = nx.from_numpy_matrix(A)
G.remove_edges_from(nx.selfloop_edges(G))

#We use the standard graph metric of degree 
#to find nodes that have at least minPts neighbors
#Degree is the number of incoming and outgoing edges a node has.
#Indegree is the number of incoming edges.
#Outdegree is the number of outgoing edges.
#In an undirected graph, Degree = Indegree = Outdegree
#In a directed graph, Degree = Indegree + Outdegree
Degree = G.degree
for node in Degree:
    idx = node[0]
    degree = node[1]
    if (degree >= minPts):
        corepoints[idx] = 1
corepoints_index_list = np.where(corepoints == 1)[0].tolist()
```

### Find Edge Points and Outliers
Here, we find all points that are reachable from any of the core points but are themselves not core points. This constitutes the set of edge points. The remaining points are considered outliers. We find all reachable points of each point not assigned to be an corepoint, using the a shortest path algorithm. This returns all the nodes that are reachable from some starting node. If this list of nodes contains at least 1 core point, our current point is considered to be an edge point. Otherwise, it is an outlier

```python
edgepoints_index_list = []
edgepoints = np.zeros((X.shape[0]))
outliers_index_list = []
outliers = np.zeros((X.shape[0]))

idx0 = np.where(corepoints == 0)[0]
for idx in idx0:
    neighbors = list(nx.shortest_path(G,idx).keys())[1:]
    if (len(np.intersect1d(corepoints_index_list,neighbors)) > 0):
        edgepoints_index_list.append(idx)
        edgepoints[idx] = 1
    else:
        outliers_index_list.append(idx)
        outliers[idx] = 1
```

### Plot

```python
plt.figure(figsize=(5,5))
plt.subplots_adjust(left=0.0,bottom=0.0, right=1.0, top=1.0, wspace=0.0, hspace=0.0)
plt.scatter(X[outliers_index_list,0],X[outliers_index_list,1],color=(0.75,0.0,0.0),s=50)
plt.scatter(X[corepoints_index_list,0],X[corepoints_index_list,1],color=(0.3,0.6,0.9),s=50)
plt.scatter(X[edgepoints_index_list,0],X[edgepoints_index_list,1],color=(0.3*0.6,0.6*0.6,0.90*0.6),s=50)
plt.grid()
plt.xlabel('x',fontsize=22,**csfont);
plt.xlabel('y',fontsize=22,**csfont);
plt.xticks(np.linspace(-2.5,2.5,9),labels=np.linspace(-2.5,2.5,9),fontsize=14,**csfont);
plt.yticks(np.linspace(-2.5,2.5,9),labels=np.linspace(-2.5,2.5,9),fontsize=14,**csfont);
plt.axis((-2.5,2.5,-2.5,2.5));
```

### Assign Clusters
After we have assigned points to be core, edge or outliers, we still need to assign them to a cluster. We can use path again, to find all neighbors of both the edge and core points. This returns a set of points (including the point itself). The number of unique sets we find represents the number of clusters. And what set an individual point is part of, determines its cluster label. By turning the list of neighbors into a tuple of neigbors, we can create a list of all tuples. Turning this list into a set removes any duplicates and return only the tuples that are different and represent all the points belonging to a single cluster.

```python
C = []
for i in range(X.shape[0]):
    if (outliers[i] == 0):
        clusterset = list(np.sort(list(nx.shortest_path(G,i).keys())))
        C.append(tuple(clusterset))
        C = list(set(C))
labels = np.zeros((X.shape[0])) - 1 
no_clusters_found = len(C)
for c,cluster in enumerate(C):
    cluster = np.array(cluster)
    labels[cluster] = c
```

#### Cluster Colors for plotting

```python
cluster_colors = np.zeros((10,3))
cluster_colors[0,:] = [1,0,0.5]
cluster_colors[1,:] = [0,0.5,0]
cluster_colors[2,:] = [1.0,0.5,0.75]
cluster_colors[3,:] = [0.25,0.75,0.25]
```

### Final Plot

```python
plt.figure(figsize=(5,5))
plt.subplots_adjust(left=0.0,bottom=0.0, right=1.0, top=1.0, wspace=0.0, hspace=0.0)
for i in range(no_clusters_found):
    idx = np.where(labels == i)[0]
    idx_edge = np.intersect1d(idx,edgepoints_index_list)
    idx_core = np.intersect1d(idx,corepoints_index_list)
    plt.scatter(X[idx_edge,0],X[idx_edge,1],color=cluster_colors[i+2,:],s=50,label='cluster ' + str(i+1) + ', edge point')
    plt.scatter(X[idx_core,0],X[idx_core,1],color=cluster_colors[i+0,:],s=50,label='cluster ' + str(i+1) + ', core point')
idx = np.where(labels == -1)[0]
plt.scatter(X[idx,0],X[idx,1],color='k',s=50,label='outlier')
leg = plt.legend(frameon=False)
plt.setp(leg.texts,fontsize=15,**csfont)
plt.xlabel('x',fontsize=22,**csfont);
plt.ylabel('y',fontsize=22,**csfont,rotation=0);
plt.xticks(np.linspace(-2.5,2.5,9),labels=np.linspace(-2.5,2.5,9),fontsize=14,**csfont);
plt.yticks(np.linspace(-2.5,2.5,9),labels=np.linspace(-2.5,2.5,9),fontsize=14,**csfont);
plt.axis((-2.5,2.5,-2.5,2.5));
plt.savefig('../../assets/figures/017/dbscan_1.png',dpi=300,bbox_inches="tight")    
```
