# Defining Clusters based on distance
### Intra/inter distance ratios
A distance based clustering method implicitely defines a cluster as a set of points that are closer to eachother, on average, compared to their distance to the nearest other cluster, or all other clusters. The following code snippet generates three clusters of data, where points are uniformly distributed around the centroids. As we push the clusters closer together and they start overlapping, we observe the ratio of between intra and interdistances asymptote to 1, indicating a total abssence of clustering for this specific definition of what constitutes a cluster. 

### Run

```python
#Parameters
no_points = 100
no_clusters = 5
no_steps = 11
distances = np.linspace(0,3,no_steps)[::-1]
cur_step = 0
cc = []
data = []

#Create Labels, displacement vectors and colors
y = []
dv = np.zeros((no_clusters,2))
cluster_colors = np.zeros((no_clusters,3))
for i in range(no_clusters):
    y = y + (np.zeros((no_points)) + i).astype(int).tolist()
    angle = i / no_clusters * 360
    dv[i,:] = [np.sin(angle/180*np.pi),np.cos(angle/180*np.pi)]
    lab = np.zeros((1,3))
    lab[0,0] = 70
    lab[0,1:] = -dv[i,:] * 60
    rgb = color.lab2rgb(lab)
    cluster_colors[i,:] = rgb
    
#Create random data within a unit circle around the origin (0,0)
r = np.sqrt(np.random.random(no_points*no_clusters))
theta = np.random.random(no_points*no_clusters) * 2 * np.pi
base = np.array([r * np.cos(theta),r * np.sin(theta)]).T
y = np.array(y)
y = np.expand_dims(y,-1)

#Create ID matrix
Ybin = np.eye(no_clusters*no_points)
for i in range(no_clusters):
    ybin = np.zeros((no_clusters*no_points,1))
    idx = np.where(y == i)[0]
    ybin[idx] = 1
    Ybin = Ybin + np.dot(ybin,ybin.T)

#Loop through a list of distances
for cur_step in range(len(distances)):

    #Move points of each cluster to their starting position
    current = base.copy()
    for i in range(no_clusters):
        idx = np.where(y == i)[0]
        current[idx,0] = current[idx,0] + (dv[i,0] * distances[cur_step])
        current[idx,1] = current[idx,1] + (dv[i,1] * distances[cur_step])
    
    #Compute Pairwise distance matrix
    data.append(current)
    D = squareform(pdist(current))
    inter = np.mean(D[Ybin == 0])
    intra = np.mean(D[Ybin == 1])
    clustering_coefficient = (inter - intra) / (intra + inter)
    cc.append(clustering_coefficient)
cc = np.array(cc)
```

### Plot

```python
plt.figure(figsize=(17,5))                                       
plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9,top=0.8, wspace=0.5, hspace=0.5)
current = data[0]
plt.subplot(1,3,1)
plt.scatter(current[:,0],current[:,1],c=cluster_colors[y,:],
            edgecolor=(0.25,0.25,0.25),linewidth=0.75,
            s=100,alpha=1.0)
plt.xlabel('x',fontsize=22,**csfont);
plt.ylabel('y',fontsize=22,**csfont);
plt.yticks(np.linspace(-4,4,5),labels=np.linspace(-4,4,5),fontsize=14,**csfont);
plt.xticks(np.linspace(-4,4,5),labels=np.linspace(-4,4,5),fontsize=14,**csfont);
plt.axis((-4,4,-4,4))
plt.grid()

current = data[-3]
plt.subplot(1,3,3)
plt.scatter(current[:,0],current[:,1],c=cluster_colors[y,:],
            edgecolor=(0.25,0.25,0.25),linewidth=0.75,
            s=100,alpha=1.0)
plt.xlabel('x',fontsize=22,**csfont);
plt.ylabel('y',fontsize=22,**csfont);
plt.yticks(np.linspace(-4,4,5),labels=np.linspace(-4,4,5),fontsize=14,**csfont);
plt.xticks(np.linspace(-4,4,5),labels=np.linspace(-4,4,5),fontsize=14,**csfont);
plt.grid()
plt.axis((-4,4,-4,4))


plt.subplot(1,3,2)
plt.plot(cc,'k')
plt.grid()
plt.scatter(0,cc[0],c='k',s=100)
plt.scatter(8,cc[-3],c='k',s=100)
plt.xlabel('Cluster Separation',fontsize=22,**csfont);
plt.ylabel('Clustering Coefficient',fontsize=22,**csfont);
plt.xticks(range(11),labels=np.round(distances,2),fontsize=14,**csfont);
plt.yticks(np.linspace(0,1,9),labels=np.linspace(0,1,9),fontsize=14,**csfont);
plt.savefig('../../assets/figures/017/cluster_definitions_1.png',dpi=300,bbox_inches="tight")
```
