### Compute Statistics Helper Function
This helper function is explained, discussed and used below

```python
def compute_stats(a,b,x,y):
    
    n = len(y)
    df = pd.DataFrame(x)
    df.columns = ['x']
    df['y'] = y
    df['x_mean'] = np.mean(x)
    df['y_mean'] = np.mean(y)
    df['y_pred'] = a*x + b
    df['(x - x_mean)^2'] = (df['x'] - df['x_mean'])**2
    df['(y - y_mean)^2'] = (df['y'] - df['y_mean'])**2
    df['(y - y_pred)^2'] = (df['y'] - df['y_pred'])**2
    df['(y_pred - y_mean)^2'] = (df['y_pred'] - df['y_mean'])**2
    df['(x - x_mean)(y - y_mean)'] = (df['x'] - df['x_mean']) * (df['y'] - df['y_mean'])

    #Calculate the statistical moments
    SST = np.sum(df['(y - y_mean)^2'].values)
    SSR = np.sum(df['(y_pred - y_mean)^2'].values)
    SSE = np.sum(df['(y - y_pred)^2'].values)
    x_var = np.sum(df['(x - x_mean)^2']) / (n-1)
    y_var = np.sum(df['(y - y_mean)^2']) / (n-1)
    x_stdev = np.sqrt(x_var)
    y_stdev = np.sqrt(y_var)
    xy_cov = np.sum(df['(x - x_mean)(y - y_mean)'])
    xy_cov = np.sum(xy_cov) / (n-1)
    xy_cor = xy_cov / np.sqrt(x_var*y_var)

    #Store
    stats = {}
    stats['x_mean'] = np.mean(x)
    stats['x_var'] = x_var
    stats['x_stdev'] = x_stdev
    stats['y_mean'] = np.mean(y)
    stats['y_var'] = y_var
    stats['y_stdev'] = y_stdev
    stats['xy_cov'] = xy_cov
    stats['xy_cor'] = xy_cor
    stats['SST'] = SST
    stats['SSR'] = SSR
    stats['SSE'] = SSE
    stats['rsquared'] = 1 - (SST/SSE)

    #Return
    return stats
```

### Examining the raw data

What ever data you might be using, it is always a good idea to plot a few summary statistics, distributions and/or individual input data points. There will always be surprises. There will always be a need for some decision making, in terms of normalizing, and because of outliers, missing values, etc. As we rely more and more on specialized toolboxes, we loose that feel for our data. Algorithms today might be black boxes, but at the very least we should know a thing or two about the data we provide it as input.

The data we use here is already a good example of what you might learn, simply by plotting various views of the data.  The original data seems to have distinct groups of data points, all of them in which medical expenses where a function of age, but at different offsets. A full exploration before subsetting is shown below. We plot medical expenses (the 'charge' column) as a function of age throughout, but color code the data point based on the additional categorical and numerical columns found in the same data set. A cursory inspection of these plots seem to suggest that being a smoker definitely increases your total health expenditures, as the majority of the data points for smokers (green dots) are shifted vertically upward (indicating higher medical expenses) relative to the non-smokers (red dots). And higher BMI does seem to lead to the same increase, although that is a bit more subtle. It does appear that high BMIs (data points colored light turqoise to yellow) on average correlate to higher medical expenses, compared to low BMIs (cyan, teal and dark blue colored data points). 

However, the number of children, sex and the geographic region of the patient does not seem to drive health cost, or at least not visibly. This leaves the three-banned structure observed in the data unexplained by the variables at our disposal. It is possible that they represent different levels of insurance coverage, or some other demographic not listed. Or, they could be an interaction effect between two variables. For example, some who smokes incurs higher costs. However, if you smoke and your a male, there is an additional increase. But not if you smoke and identify as female. In otherwords, your gender only influences your medical cost if you happen to be a smoker. We invite you to look for whether interactions effects in this data can explain it, and we would love to hear what you find. 

```python
#Plot of the original data, with medical expenses plotted as a function of age
plt.figure(figsize=(9,8))
plt.subplot(2,2,1)
plt.scatter(x,y,c='k')
plt.grid()
plt.xticks(range(10,80,10),labels=[])
plt.yticks(range(0,100000,10000),labels=range(0,100000,10000),fontsize=14,**csfont);
plt.title('all',fontsize=20,**csfont);  
plt.ylabel('medical expenses (dollars)',fontsize=18,**csfont);

#Colormap relatred calls
cmap = mpl.cm.viridis
colormap = np.array(cmap.colors)
categorical_colors = [
    [1.0,0.0,0.0],
    [0.0,0.5,0.0],
    [0.2,0.5,0.6],
    [0.6,0.2,0.2]]

#Color coding the data as a function of three categorical variables
categorical_columns = ['sex','smoker','region']
for c,column in enumerate(categorical_columns):
    plt.subplot(2,2,c+2)
    plt.grid()
    z = data[column].values
    classes = np.unique(z)
    for i in range(len(classes)):
        ind = np.where(z == classes[i])[0]
        x_subset = x[ind]
        y_subset = y[ind]
        plt.scatter(x_subset,y_subset,color=categorical_colors[i])
        if (c+2 == 2):         
            plt.xticks(range(10,80,10),labels=[])
            plt.yticks(range(0,100000,10000),labels=[])
        if (c+2 == 3):         
            plt.xticks(range(10,80,10),[],fontsize=14,**csfont);
            plt.yticks(range(0,100000,10000),labels=range(0,100000,10000),fontsize=14,**csfont);        
            plt.ylabel('medical expenses (dollars)',fontsize=18,**csfont);
        if (c+2 == 4):
            plt.xticks(range(10,80,10),labels=[])
            plt.yticks(range(0,100000,10000),labels=[])
    plt.legend(classes,frameon=False)
    plt.title(column,fontsize=20,**csfont);  
plt.savefig('../../assets/figures/014/data_exploration_categorical.png',bbox_inches="tight")        
    

#Color coding the data as a function of two numerical variables
plt.figure(figsize=(11.25,4))
numerical_columns = ['bmi','children']
for c,column in enumerate(numerical_columns):
    plt.subplot(1,2,1+c)
    plt.grid()
    z = data[column].values
    minval = np.min(z)
    maxval = np.max(z)
    z = z - minval
    z = z / (maxval - minval)
    z = (z * 255).astype(int)
    colors = colormap[z,:]
    plt.scatter(x,y,c=colors)
    if (c == 0): 
        plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=14,**csfont);
        plt.yticks(range(0,80000,8000),labels=range(0,80000,8000),fontsize=14,**csfont);        
        plt.xlabel('age',fontsize=18,**csfont);   
        plt.ylabel('medical expenses (dollars)',fontsize=18,**csfont);
    if (c == 1): 
        plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=14,**csfont);
        plt.yticks(range(0,80000,8000),labels=[],fontsize=14,**csfont);        
        plt.xlabel('age',fontsize=18,**csfont);   
    
    plt.title(column,fontsize=20,**csfont); 
    colorbar = plt.colorbar()  
    minval = np.floor(minval)
    maxval = np.ceil(maxval)
    crange = np.round(np.linspace(0,1,11),2)
    colorbar.set_ticks(crange)
    colorbar.set_ticklabels(minval + crange*maxval)
    colorbar.set_label(column,fontsize=14,**csfont);        
plt.savefig('../../assets/figures/014/data_exploration_numerical.png',bbox_inches="tight")    
```

### Clean and subsampling the data
Given the findings above, I've taken the liberty to take out some data points (see below). We have deliberatly removed the two upper, more noise bands that seem to form when we plot medical expenses as a function of age. In addition, we transformed the medical expenses data by dividing it by 1000, so that values represent dollar x 1000. In doing so, we create a clean data set, which shows a surprisingly clean relationship, ideal for trying our hands with regression for the first time. You might argue that we obtained this clean subset  by selectively removing data points. However, even though we are cherry picking somewhat, we only excluded data points based on some absolute threshold. 

```python
ind0 = np.where(y_orig > 17000)[0]
ind1 = np.where(x_orig < 46)[0]
ind2 = np.where(y_orig > 10000)[0]
ind3 = np.intersect1d(ind1,ind2)
ind4 = np.union1d(ind3,ind0)
ind5 = range(x_orig.shape[0])
ind5 = np.setdiff1d(ind5,ind4)
x = x_orig[ind5].astype('float')
y = y_orig[ind5].astype('float')
y = y / 1000
```

# Our first manually created regression model
Our first model will simply output the mean medical expense across all of our data. In doing so, we implicitely create a model that per definition will not explain any of the observed variance in y. We can derive this from the equations for the sum of squares total (SST), sum of squares regression (SSR) and sum of squares error (SSE) below. For now, our first model is defined as follows:

&nbsp;&nbsp;&nbsp;&nbsp;$a = 0, b = \bar{y}$<br>
&nbsp;&nbsp;&nbsp;&nbsp;$\hat{y} = ax + b = \bar{y}$

### Compute all statistics, given parameters a and b, input data x, and output data y.
We illustration purposes, we will use a pandas dataframe to compute all the necessary intermediate values that will give us the mean statistical metric we need. The helper function that does this is defined above 

```python
stats =  compute_stats(0,np.mean(y),x,y)
print(stats)
```

```python
# x = np.array([25, 27, 29])
# y = np.array([5, 15, 9])
x_var = (x - np.mean(x)) **2
y_var = (y - np.mean(y)) **2
x_var = np.sum(x_var) / (len(x)-1)
y_var = np.sum(y_var) / (len(y)-1)
print(x_var)
print(y_var)
# xy_cov = (x - np.mean(x)) * (y - np.mean(y))
# xy_cov = np.sum(xy_cov) / (len(x)-1)
# xy_cov / np.sqrt(x_var*y_var)
```

### Plot Data and Model

```python
a = 0
b = np.mean(y)
w = [a,b]
ypred = x*a + b

#Left and right point of the linear regression line
x0 = 10
x1 = 70
y0 = x0*a + b
y1 = x1*a + b

SST = np.sum((y - np.mean(y))**2)
SSE = np.sum((y - ypred)**2)
SSR = np.sum((ypred - np.mean(y))**2)
rsquare = 1 - (SSE/SST)

print(SST)
print(SSE)
print(SSR)
print(rsquare)

plt.figure(figsize=(6,6));
plt.scatter(x,y,c='k');
plt.plot([x0,x1],[y0,y1],'k',zorder=100);
plotrange = list(range(x.shape[0]))
for index in plotrange:
    plt.plot([x[index],x[index]],[y[index],ypred[index]],'r--')
    plt.scatter(x[index],ypred[index],c='red',zorder=100);
plt.grid();
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.axis((10,70,0,15));
plt.xlabel('age',fontsize=22,**csfont);
plt.ylabel('medical expenses (dollars x 1000)',fontsize=22,**csfont);
plt.savefig('../../assets/figures/014/rsquare_01.png', bbox_inches="tight")
```
