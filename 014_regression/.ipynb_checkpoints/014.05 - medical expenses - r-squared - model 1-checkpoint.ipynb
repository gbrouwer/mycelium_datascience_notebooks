# Manual Model 1
Our first model just predict the mean value of y

### Parameters

```python
a = 0
b = 0
```

### Import

```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

### Compute the Statistics

```python
def compute_stats(a,b,x,y):
    
    n = len(y)
    df = pd.DataFrame(x)
    df.columns = ['x']
    df['y'] = y
    df['x_mean'] = np.mean(x)
    df['y_mean'] = np.mean(y)
    df['y_pred'] = a*x + b
    df['(x - x_mean)^2'] = (df['x'] - df['x_mean'])**2
    df['(y - y_mean)^2'] = (df['y'] - df['y_mean'])**2
    df['(y - y_pred)^2'] = (df['y'] - df['y_pred'])**2
    df['(y_pred - y_mean)^2'] = (df['y_pred'] - df['y_mean'])**2
    df['(x - x_mean)(y - y_mean)'] = (df['x'] - df['x_mean']) * (df['y'] - df['y_mean'])

    #Calculate the statistical moments
    SST = np.sum(df['(y - y_mean)^2'].values)
    SSR = np.sum(df['(y_pred - y_mean)^2'].values)
    SSE = np.sum(df['(y - y_pred)^2'].values)
    x_var = np.sum(df['(x - x_mean)^2']) / (n-1)
    y_var = np.sum(df['(y - y_mean)^2']) / (n-1)
    x_stdev = np.sqrt(x_var)
    y_stdev = np.sqrt(y_var)
    xy_cov = np.sum(df['(x - x_mean)(y - y_mean)'])
    xy_cov = np.sum(xy_cov) / (n-1)
    xy_cor = xy_cov / np.sqrt(x_var*y_var)

    #Store
    stats = {}
    stats['x_mean'] = np.mean(x)
    stats['x_var'] = x_var
    stats['x_stdev'] = x_stdev
    stats['y_mean'] = np.mean(y)
    stats['y_var'] = y_var
    stats['y_stdev'] = y_stdev
    stats['xy_cov'] = xy_cov
    stats['xy_cor'] = xy_cor
    stats['SST'] = SST
    stats['SSR'] = SSR
    stats['SSE'] = SSE
    stats['rsquared'] = 1 - (SSE/SST)

    #Return
    return stats,df
```

### Load Data

```python
data = pd.read_csv('../../data/healthcare_cost/insurance.csv')
columns = data.columns
x_orig = data.iloc[:,0].values
y_orig = data.iloc[:,-1].values
```

### Clean Dataset for the univariate example

```python
ind0 = np.where(y_orig > 17000)[0]
ind1 = np.where(x_orig < 46)[0]
ind2 = np.where(y_orig > 10000)[0]
ind3 = np.intersect1d(ind1,ind2)
ind4 = np.union1d(ind3,ind0)
ind5 = range(x_orig.shape[0])
ind5 = np.setdiff1d(ind5,ind4)
x_orig = x_orig[ind5].astype('float')
y_orig = y_orig[ind5].astype('float')
```

### Compute all Three Sum of Squares and R-square

```python
x0 = 10
x1 = 70
y0 = np.mean(y)
y1 = np.mean(y)
y_pred = y.copy()
y_pred[:] = np.mean(y)
y_mean = np.mean(y)
SST = np.sum((y - np.mean(y))**2)
SSR = np.sum((y_pred - np.mean(y))**2)
SSE = np.sum((y - y_pred)**2)
rsquared = SSR/SST
```

### Show Contributions of SSR, SST and SSE
