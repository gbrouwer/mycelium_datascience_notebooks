### Linear Regression - Statistical Approach - Breast Cancer Dataset
This demo will walk through the algebriac approach to linear regression, computing the best fitting linear function of y = ax + b. To do so we will use the Wisconscin Breast Cancer Dataset. Since this is a categorical dataset in nature (the outcome represents whether the tumor is malignant or benign), we instead use two input features as input and output. Specifically, we will be using the feature radius_mean to predict both concave_points_mean and symmetry_mean.

### Import

```python
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load Data
This loads both the diagnostic and prognostic data sets from Wisconsin Breat Cancer Dataset. 
The dataset used in this story is publicly available and was created by Dr. William H. Wolberg, physician at the University Of Wisconsin Hospital at Madison, Wisconsin, USA. To create the dataset Dr. Wolberg used fluid samples, taken from patients with solid breast masses. The samples were processed through an automated process, and ten features from each one of the cells in the sample were computed. Averaging across all cells in the sample, it returned a mean value, extreme value and standard error of each feature, returning a 30 real-valuated vector.A full detailed explanation of what this data contains can be found here: http://archive.ics.uci.edu/ml/datasets/breast+cancer+wisconsin+%28diagnostic%29

```python
columns = []
with open('../../data/breast_cancer/wdbc.columns','r') as f:
    for line in f:
        columns.append(line.rstrip())
data = pd.read_csv('../../data/breast_cancer/wdbc.data',header=None)
data.columns = columns
```

### Split the data into input X and output y and drop the ID

```python
X = data[columns[2:]].values
y = data[columns[1]].values
```

### Recode y into a numerical variable

```python
classes = np.unique(y)
for i,item in enumerate(classes):
    indices = np.where(y == item)[0]
    y[indices] = i
```

### Select input and output data

```python
x = X[:,2] #radius_mean
y = X[:,12] #concave_mean
```

### Plot the data
Plotting the data shows a very strong correlation between variables. This should yield a reasonable fit. 

```python
plt.figure(figsize=(6,6));
plt.scatter(x,y,c='k');
plt.axis((0,200,0,10))
plt.grid()
```

### We use the formula for ordinary least squares to compute a and b
Note that we first add a column of 1 to the input data x. This extra colunm allows us to compute both the slope (a) and intercept (b). Also, we need to make sure X is a matrix of nx2 (n datapoints and one column for the radius_mean and one column of ones for the intercept) and y is a column vector of nx1 (n datapoints with one column for the 'concave_mean' data). 

```python
x = np.vstack((x,np.ones(x.shape))).T
y = np.expand_dims(y,-1)
```

### Ordinary Least Squares

```python
w = np.dot(np.dot(np.linalg.inv(np.dot(x.T,x)),x.T),y)
a = w[0]
b = w[1]
```

### Plot the data overlayed with the fitted function
The fit is identical to the statistical method, as expected. The formulas might look very different, they compute the same thing from the same data. 

```python
xval = np.linspace(np.min(x[:,0]),np.max(x[:,0]),100)
yval = xval*a + b;
plt.figure(figsize=(6,6));
plt.axis((0,200,0,10))
plt.grid()
plt.scatter(x[:,0],y,c='k');
plt.plot(xval,yval,'r',linewidth=2);
```

### R-squared

To evaluate the quality of our fit, we compute the R-square. A value between 0 and 1, values close to 1 indicate a good fit of the model. Please note that R-square is a different metric than correlation or the slope of the fitted function. In the absence of a strong correlation between x and y, we can still have a good fit. The opposite is true as well: a strong correlation might still yield a poor fit if our model can't capture the true nature of the relationship between x and y. For example, the relationship might be highly nonlinear, so a linear model will not be able to capture that.  

To compute the R-square, we need to compute the SSR (sum of squares residual, or the amount of variance left over after we fit the function) and the SST (the total amount of variance in the data). 

```python
SSR = np.sum((y - (a*x + b))**2)
SST = np.sum( (y-np.mean(y[:,0]))**2)
rsquared = 1 - (SSR/SST)
print(SSR)
print(SST)
print(rsquared)
```
