### Linear Regression - Statistical Approach - Breast Cancer Dataset
This demo will walk through the statistical approach to linear regression, computing the best fitting linear function of y = ax + b. To do so we will use the Wisconscin Breast Cancer Dataset. Since this is a categorical dataset in nature (the outcome represents whether the tumor is malignant or benign), we instead use two input features as input and output. Specifically, we will be using the feature radius_mean to predict both concave_points_mean and symmetry_mean

### Import

```python
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Load Data
This loads both the diagnostic and prognostic data sets from Wisconsin Breat Cancer Dataset. 
The dataset used in this story is publicly available and was created by Dr. William H. Wolberg, physician at the University Of Wisconsin Hospital at Madison, Wisconsin, USA. To create the dataset Dr. Wolberg used fluid samples, taken from patients with solid breast masses. The samples were processed through an automated process, and ten features from each one of the cells in the sample were computed. Averaging across all cells in the sample, it returned a mean value, extreme value and standard error of each feature, returning a 30 real-valuated vector.A full detailed explanation of what this data contains can be found here: http://archive.ics.uci.edu/ml/datasets/breast+cancer+wisconsin+%28diagnostic%29

```python
columns = []
with open('../../data/breast_cancer/wdbc.columns','r') as f:
    for line in f:
        columns.append(line.rstrip())
data = pd.read_csv('../../data/breast_cancer/wdbc.data',header=None)
data.columns = columns
```

### Split the data into input X and output y and drop the ID

```python
X = data[columns[2:]].values
y = data[columns[1]].values
```

### Recode y into a numerical variable

```python
classes = np.unique(y)
for i,item in enumerate(classes):
    indices = np.where(y == item)[0]
    y[indices] = i
```

### Visualize the distribution of all variables

```python
plt.figure(figsize=(12,12))
count = 0
for i in range(6):
    for j in range(5):
        plt.subplot(6,5,count+1)
        plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)        
        x = X[:,i]
        hist = np.histogram(x,20)
        c1 = i / 6
        c2 = j / 5
        plt.bar(range(len(hist[0])),hist[0],color=(c1,0.5,c2,1))
        if (i == 5 or j==0):
            if (i == 5 and j == 0):
                plt.xticks(fontsize=14,**csfont)
                plt.yticks(fontsize=14,**csfont)
                plt.xlabel('bins',fontsize=20,**csfont);
                plt.ylabel('counts',fontsize=20,**csfont);
            if (j > 0):
                plt.yticks([])
                plt.xlabel('bins',fontsize=20,**csfont);
            if (i < 5):
                plt.xticks([])
                plt.ylabel('counts',fontsize=20,**csfont);
        else:
            plt.xticks([])
            plt.yticks([])
        plt.title(columns[count+2],**csfont,fontsize=14)
        count = count + 1
plt.savefig('../../assets/figures/014/statistical_approach_breastcancer_01.png', bbox_inches="tight")        
```

### Select input and output data

```python
x = X[:,2] #radius_mean
y = X[:,10] #concave_mean
```

### Plot the data
Plotting the data shows a very strong correlation between variables. This should yield a reasonable fit. 

```python
plt.figure(figsize=(6,6));
plt.scatter(x,y,c='k');
plt.grid()
plt.axis((0,200,0,1.5));
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel('radius (mean)',fontsize=20,**csfont);
plt.ylabel('concave (mean)',fontsize=20,**csfont);
plt.savefig('../../assets/figures/014/statistical_approach_breastcancer_02.png', bbox_inches="tight");
```

### Computing the mean

```python
x_mean = np.sum(x) / len(x)
y_mean = np.sum(y) / len(y)
```

### Compute the sum of squares for x and y

```python
x_sum_of_squares = np.sum((x - x_mean)**2)
y_sum_of_squares = np.sum((y - y_mean)**2)
```

### Compute the sum of squares between x and y

```python
xy_sum_of_squares = np.sum(((x-x_mean)*(y-y_mean)))
```

### Compute 'a' aka the slope of the function

```python
a = xy_sum_of_squares / x_sum_of_squares
```

### Compute 'b' aka the intercept of the function

```python
b = y_mean - x_mean*a
```

### Plot the data overlayed with the fitted function
The fit is identical to the algebriac (OLS) method, as expected. The formulas might look very different, they compute the same thing from the same data.

```python
xval = np.linspace(np.min(x),np.max(x),100)
yval = xval*a + b;
plt.figure(figsize=(6,6));
plt.plot(xval,yval,'r',linewidth=3);
plt.scatter(x,y,c='k');
plt.grid()
plt.axis((0,200,0,1.5));
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel('radius (mean)',fontsize=20,**csfont);
plt.ylabel('concave (mean)',fontsize=20,**csfont);
plt.savefig('../../assets/figures/014/statistical_approach_breastcancer_03.png', bbox_inches="tight");
print(a)
print(b)
```

### R-squared

To evaluate the quality of our fit, we compute the R-square. A value between 0 and 1, values close to 1 indicate a good fit of the model. Please note that R-square is a different metric than correlation or the slope of the fitted function. In the absence of a strong correlation between x and y, we can still have a good fit. The opposite is true as well: a strong correlation might still yield a poor fit if our model can't capture the true nature of the relationship between x and y. For example, the relationship might be highly nonlinear, so a linear model will not be able to capture that.  

To compute the R-square, we need to compute the SSR (sum of squares residual, or the amount of variance left over after we fit the function) and the SST (the total amount of variance in the data). 

```python
SSR = np.sum((y - (a*x + b))**2)
SST = np.sum((y-y_mean)**2)
rsquared = 1 - (SSR/SST)
print(rsquared)
```

### Let's do that again, but now radius_mean vs symmetry_mean

```python
x = X[:,2] #radius_mean
y = X[:,12] #symmetry_mean
```

### Recompute

```python
x_mean = np.sum(x) / len(x)
y_mean = np.sum(y) / len(y)
x_sum_of_squares = np.sum((x - x_mean)**2)
y_sum_of_squares = np.sum((y - y_mean)**2)
xy_sum_of_squares = np.sum(((x-x_mean)*(y-y_mean)))
a = xy_sum_of_squares / x_sum_of_squares
b = y_mean - x_mean*a
xval = np.linspace(np.min(x),np.max(x),100)
yval = xval*a + b;
```

### Plot the result
Seems now our perception and the resulting fit match. There doesn't appear to be much correlation, and indeed the resulting fit is flat. 

```python
xval = np.linspace(np.min(x),np.max(x),100)
yval = xval*a + b;
plt.figure(figsize=(6,6));
plt.plot(xval,yval,'r',linewidth=3);
plt.scatter(x,y,c='k');
plt.grid()
plt.axis((0,200,0,12));
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel('radius (mean)',fontsize=20,**csfont);
plt.ylabel('concave (mean)',fontsize=20,**csfont);
plt.savefig('../../assets/figures/014/statistical_approach_breastcancer_04.png', bbox_inches="tight");
print(a)
print(b)
```

### This poor fit should be reflected in the rsquare

```python
SSR = np.sum((y - (a*x + b))**2)
SST = np.sum((y-y_mean)**2)
rsquared = 1 - (SSR/SST)
print(rsquared)
```
