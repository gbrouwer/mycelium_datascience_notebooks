### Import

```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

### Fonts
Define fonts to use plot fonts that are similar to the classic LaTEX fonts used for the equations

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Load Data
This data is a subset of a dataset hosted on [Kaggle](https://www.kaggle.com/datasets/mirichoi0218/insurance)

```python
data = pd.read_csv('../../data/healthcare_cost/insurance.csv')
columns = data.columns
x = data.iloc[:,0].values
y = data.iloc[:,-1].values
```

### Clean Dataset for the univariate example
I've taken the liberty to take out some data points. The original data seems to have distinct groups of data points, all of them in which medical expenses where a function of age, but at different offsets. A full exploration before subsetting is shown below. We plot medical expenses (charge column) as a function of age throughout, but color code the data point based on the additional categorical and numerical columns found in the same data set. A cursory inspection of these plots seem to suggest that being a smoker definitely increases your total health expenditures. And higher BMI does seem to lead to the same increase. However, the number of children, sex and the region of the patient does not seem to drive health cost that much. This leaves the three ban structure observed in the data unexplained by the variables at our disposal. It is possible that they represent different levels of insurance coverage, or some other demographic not listed. 

```python
plt.figure(figsize=(14,14))
plt.subplot(3,3,1)
plt.scatter(x,y,c='k')
plt.grid()
plt.xticks(range(10,80,10),labels=[])
plt.yticks(range(0,100000,10000),labels=range(0,100000,10000),fontsize=14,**csfont);
plt.ylabel('medical expenses (dollars)',fontsize=16,**csfont);
plt.title('all',fontsize=20,**csfont);  

categorical_colors = [
    [1.0,0.0,0.0],
    [0.0,0.5,0.0],
    [0.2,0.5,0.6],
    [0.6,0.2,0.2]]

categorical_columns = ['sex','smoker','region']
numerical_columns = ['bmi','children']
for c,column in enumerate(categorical_columns):
    plt.subplot(3,3,c+2)
    plt.grid()
    z = data[column].values
    classes = np.unique(z)
    for i in range(len(classes)):
        ind = np.where(z == classes[i])[0]
        x_subset = x[ind]
        y_subset = y[ind]
        plt.scatter(x_subset,y_subset,color=categorical_colors[i])
        if (c+2 < 4):         
            plt.xticks(range(10,80,10),labels=[])
            plt.yticks(range(0,100000,10000),labels=[])

    if (c+2 == 4):
        plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=14,**csfont);
        plt.yticks(range(0,100000,10000),labels=range(0,100000,10000),fontsize=14,**csfont);        
        plt.xlabel('age',fontsize=16,**csfont);
        plt.ylabel('medical expenses (dollars)',fontsize=16,**csfont);
        
    plt.legend(classes,frameon=False)
    plt.title(column,fontsize=20,**csfont);  
    
for c,column in enumerate(numerical_columns):
    plt.subplot(3,3,c+2+len(categorical_columns))
    plt.grid()
    z = data[column].values
    minval = np.min(z)
    maxval = np.max(z)
    z = z - minval
    z = z / (maxval - minval)
    colors = np.zeros((len(x),3))
    colors[:,0] = z**4
    colors[:,1] = (1-z)**2
    colors[:,2] = z    
    plt.scatter(x,y,c=colors)
    plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=14,**csfont);
    plt.yticks(range(0,100000,10000),labels=[],fontsize=14,**csfont);
    plt.xlabel('age',fontsize=16,**csfont);   
    plt.title(column,fontsize=20,**csfont);  
plt.savefig('../../assets/figures/014/data_exploration.png',bbox_inches="tight")    
```

### Subsetting the data
Here we deliberatly remove the two upper, more noise bands that seem to form when we plot medical expenses as a function of age. In addition, we will transform the medical expenses data by dividing it by 1000, so that values represent dollar x 1000. In doing so, we create a suprisingly clean data set. Ideal for trying our hands with regression for the first time

```python
ind0 = np.where(y > 17000)[0]
ind1 = np.where(x < 46)[0]
ind2 = np.where(y > 10000)[0]
ind3 = np.intersect1d(ind1,ind2)
ind4 = np.union1d(ind3,ind0)
ind5 = range(x.shape[0])
ind5 = np.setdiff1d(ind5,ind4)
x = x[ind5].astype('float')
y = y[ind5].astype('float')
```

### Plot Full

```python
stepsize = 10
plt.figure(figsize=(6,6));
plt.scatter(x,y,c='k');
plt.grid();
plt.xlabel('age',fontsize=22,**csfont);
plt.ylabel('medical expenses (dollars)',fontsize=22,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.axis((10,70,0,20000));
plt.savefig('../../assets/figures/014/statistical_approach_full.png',bbox_inches="tight")
```

### Random Subset
To follow along with all necessary computation, we restrict the data to a mere 20 points. Luckily, the data has such a strong effect of age on medical expenses that we can safely subset without risking finding substantially different paramters for our best-fitting regression function

```python
np.random.seed(100)
idx = np.random.randint(0,x.shape[0],20)
x = x[idx]
y = y[idx]
meany = np.mean(y)
```

### Statistical Method

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
ss_x = np.sum((x - np.mean(x))**2)
ss_y = np.sum((y - np.mean(y))**2)
ss_xy = np.sum((x - np.mean(x))*(y - np.mean(y)))
a = ss_xy / ss_x
b = y_mean - (a * x_mean)
print(a,b)
```

### Predict and compute the SSE, SST, SSR, and r-square

```python
ypred = a*x + b
x0 = 15
x1 = 65
y0 = a*x0 + b
y1 = a*x1 + b

SST = np.sum((y - np.mean(y))**2)
SSE = np.sum((y - ypred)**2)
SSR = np.sum((ypred - np.mean(y))**2)
rsquare = 1 - (SSE/SST)
print(SST)
print(SSE)
print(SSR)
print(rsquare)
```

### Plot the result

```python
stepsize = 10
plt.figure(figsize=(6,6));
plt.scatter(x,y,c='k');
plt.plot([x0,x1],[y0,y1],'k',zorder=100)

plotrange = list(range(x.shape[0]))
for index in plotrange:
    plt.plot([x[index],x[index]],[y[index],ypred[index]],'r--')
    plt.scatter(x[index],ypred[index],c='red',zorder=100);

plt.grid();
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel('age',fontsize=22,**csfont);
plt.ylabel('medical expenses (dollars)',fontsize=22,**csfont)
plt.axis((10,70,0,15000));
plt.savefig('../../assets/figures/014/statistical_approach_subset_fit.png', bbox_inches="tight")
```
