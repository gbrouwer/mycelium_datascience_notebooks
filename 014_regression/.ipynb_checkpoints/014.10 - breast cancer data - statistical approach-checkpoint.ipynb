### Mean

```python
def mean(x):

    n = len(x)
    return np.sum(x) / n
```

### Variance

```python
def variance(x,sampleType):
    
    n = len(x)
    mean_x = mean(x)
    if (sampleType == 'population'):
        return np.sum((mean_x - x)**2) / n
    else:
        return np.sum((mean_x - x)**2) / (n-1)
    
```

### Standard Deviation

```python
def std_deviation(x,sampleType):

    return np.sqrt(variance(x,sampleType))
```

### Covariance

```python
def covariance(x,y,sampleType):
   
    n = len(x)
    if (sampleType == 'population'):
        return np.sum((x - mean(x)) * (y - mean(y))) / (n-1)

    else:
        return  np.sum((x - mean(x)) * (y - mean(y))) / n
```

### Select input and output data. We start with radius and compactness. In other words, can we predict the radius if we knew the compactness of a tumor?

```python
colx = 0
coly = 5
x = X[:,colx]
y = X[:,coly]
print(columns[colx])
print(columns[coly])
```

### Plot the data
Plotting the data shows a very strong correlation between variables. This should yield a reasonable fit. 

```python
plt.figure(figsize=(6,6));
plt.scatter(x,y,c='k');
plt.grid()
plt.axis((np.min(x),np.max(x),np.min(y),np.max(y)))
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel(columns[colx],fontsize=20,**csfont);
plt.ylabel(columns[coly],fontsize=20,**csfont);
plt.savefig('../../assets/figures/014/statistical_approach_breastcancer_02.png', bbox_inches="tight");
```

### Computing the mean

```python
x_mean = np.sum(x) / len(x)
y_mean = np.sum(y) / len(y)
```

### Compute the sum of squares for x and y

```python
x_sum_of_squares = np.sum((x - x_mean)**2)
y_sum_of_squares = np.sum((y - y_mean)**2)
```

### Compute the sum of squares between x and y

```python
xy_sum_of_squares = np.sum(((x-x_mean)*(y-y_mean)))
```

### Compute 'a' aka the slope of the function

```python
a = xy_sum_of_squares / x_sum_of_squares
```

### Compute 'b' aka the intercept of the function

```python
b = y_mean - x_mean*a
```

### Plot the data overlayed with the fitted function

```python
xval = np.linspace(np.min(x),np.max(x),100)
yval = xval*a + b;
plt.figure(figsize=(6,6));
plt.plot(xval,yval,'r',linewidth=3);
plt.scatter(x,y,color=(0.0,0.0,0.0),s=35)
plt.grid()
plt.axis((np.min(x),np.max(x),np.min(y),np.max(y)))
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel(columns[colx],fontsize=20,**csfont);
plt.ylabel(columns[coly],fontsize=20,**csfont);
plt.savefig('../../assets/figures/014/statistical_approach_breastcancer_03.png', bbox_inches="tight");
```

### R-squared

To evaluate the quality of our fit, we compute the R-square. As we discussed previously, the r-squared is a value between 0 and 1, values close to 1 indicate a good fit of the model. Please note that R-square is a different metric than correlation or the slope of the fitted function. In the absence of a strong correlation between x and y, we can still have a good fit. The opposite is true as well: a strong correlation might still yield a poor fit if our model can't capture the true nature of the relationship between x and y. For example, the relationship might be highly nonlinear, so a linear model will not be able to capture that.  

To compute the R-square, we need to compute the SSR (sum of squares residual, or the amount of variance left over after we fit the function) and the SST (the total amount of variance in the data). 

```python
SSR = np.sum((y - (a*x + b))**2)
SST = np.sum((y-y_mean)**2)
rsquared = 1 - (SSR/SST)
print(rsquared)
```

### Let's do that again, but now radius_mean vs area_mean

```python
colx = 0
coly = 3
x = X[:,colx]
y = X[:,coly]
print(columns[colx])
print(columns[coly])
```

### Recompute

```python
x_mean = np.sum(x) / len(x)
y_mean = np.sum(y) / len(y)
x_sum_of_squares = np.sum((x - x_mean)**2)
y_sum_of_squares = np.sum((y - y_mean)**2)
xy_sum_of_squares = np.sum(((x-x_mean)*(y-y_mean)))
a = xy_sum_of_squares / x_sum_of_squares
b = y_mean - x_mean*a
xval = np.linspace(np.min(x),np.max(x),100)
yval = xval*a + b;
```

### Plot the result
When we carefullt consider the variables in the columns, we should realize that the radius, area and perimeter of a 2D slice through a tumor wil be highly correlated. Nature favors rounded shapes, so the outline of a tumor is likely to be circular to some degree. And for a perfect circle, the relationship between area, radius and perimeter is fixed. Therefore, for structures that are circular, albeit not perfectly, the radius, area and perimeter should be highly correlated. This is indeed what we see below when we look at the radius versus area variables. Such highly correlated variables in our input data can be a significant problem for our machine learning algorithm. Mathematically, a matrix of input variables/regressors of which some are interdependent (i.e. statistically related, or correlated) can be ill-conditioned. Without getting to technical, an ill-conditioned matrix is a matrix that has too many dimensions for what it is trying to explain. Area, radius and perimeter are so highly correlated that mathemtically they can fully captured by a single dimension, rather than three, of which two are redudant. An ill-conditioned matrix cannot be properly inversed, and has no unique solutions. In other words, our algorithm doesn't know how to distribute the variance in the data across the regressors. Does it put it all on one regressor (e.g. area) or does it spread it out across all three?<br>

This example provides some useful guidance in exploring your input data. Try to take the mathematical inverse of at least a portion of your total dataset. Should your tool of choice (e.g. numpy) complain, take a closer look. Typically, the solution to this problem involves ommitted the redudant variables. If the interdependence isn't strong enough to cause problems during inversion, consider running a dimensionality reduction algorithm (see Chapter 17). And keep in mind that dependending on the tool you are using, you might not get notified your matrices are indeed ill-conditioned. For example, Tensorflow might just happily try and fit your data, even though it truly can't, mathematically speaking. 

```python
xval = np.linspace(np.min(x),np.max(x),100)
yval = xval*a + b;
plt.figure(figsize=(6,6));
plt.plot(xval,yval,'r',linewidth=3);
plt.scatter(x,y,c='k');
plt.grid()
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel(columns[colx],fontsize=20,**csfont);
plt.ylabel(columns[coly],fontsize=20,**csfont);
plt.savefig('../../assets/figures/014/statistical_approach_breastcancer_04.png', bbox_inches="tight");
```

### Relationship between Radius and Area
The relationship between area and radius is given by area = radius*a + b. Is this inline with our assumption that tumors, when viewed in 2D biopsy slices, are largely circular? We know that the area of a circle is:

$area = \pi \cdot radius^2$

If we pick a radius of 20, we thus expect an area of:

$area = \pi \cdot 400 = 3.14159275 \cdot 400 = 1257$

Which is very similar to our prediction from the model we trained:

$area = (20 \cdot a) + b = (20 \cdot 98.60) - 738 = 1234$

### This good fit should be reflected in the rsquare

```python
SSR = np.sum((y - (a*x + b))**2)
SST = np.sum((y-y_mean)**2)
rsquared = 1 - (SSR/SST)
print(rsquared)
```
