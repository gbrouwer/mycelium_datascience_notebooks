### Add Column of ones

```python
x_norm = np.expand_dims(x_norm,-1)
o = np.ones((x_norm.shape[0],1))
X_norm = np.hstack((x_norm,o))
```

### Divide Data

```python
X_train = X_norm[:600,:].copy()
X_test = X_norm[600:,:].copy()
y_train = y_norm[:600].copy()
y_test = y_norm[600:].copy()
```

### Init

```python
learning_rate = 0.02
w = np.random.randn((2))
batch_size = 1
no_epoch = 1000
```

### Gradient Descent

```python
w_history = []
loss_history = []
for epoch in range(no_epoch):
    
    randval = np.random.permutation(X_train.shape[0])
    X_batch = X_train[:batch_size,:]
    y_batch = y_train[:batch_size]
    y_pred = np.dot(X_batch,w)
    E = (y_batch - y_pred)
    loss = np.sum(E**2)
    loss_history.append(loss)
    w_history.append(w)

    delta = np.dot(X_batch.T,E) * (-2/X_batch.shape[0])
    delta = delta * learning_rate
    w = w - delta
loss_history = np.array(loss_history)
```

### Plot Loss

```python
plt.figure(figsize=(6,6));
plt.plot(loss_history)
plt.grid();
plt.xlabel('Epoch',fontsize=18);
plt.ylabel('Loss',fontsize=18);
```

### Create Regression Line

```python
x0 = 0.25
x1 = 1.75
y0 = w[0]*x0 + w[1]
y1 = w[0]*x1 + w[1]
```

### Plot Loss Field

```python
aval = np.linspace(-2,2,21)
bval = np.linspace(-2,2,21)
Xm,Ym = np.meshgrid(aval, bval)
XX = np.array([Xm.ravel(), Ym.ravel()]).T
w_history = np.array(w_history)
loss_history = np.array(loss_history)
lossfield = np.zeros((21,21)).T
for i in range(len(aval)):
    for j in range(len(bval)):
        w = np.array([aval[i],bval[j]])
        w = np.expand_dims(w,-1)
        y_pred = np.dot(X_batch,w)
        E = (y_batch - y_pred)
        loss = np.sum(E**2) / X_batch.shape[0]
        lossfield[j,i] = loss
plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(Xm,Ym,lossfield,cmap='YlGnBu', edgecolor='gray')
ax.set_title('Surface plot')
ax.view_init(elev=30., azim=310)
ax.plot(w_history[:,0],w_history[:,1],loss_history,zorder=100,c='k')
plt.axis((-2,2,-2,2))
plt.savefig('../../assets/figures/test.png');
plt.show();    
```
