# Medical Expense Dataset - Data exploration
This notebook explores the medical expense dataset, which we will use moving forward to demo various ways of estimating linear regression models. 

### Import

```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from PIL import Image
from mpl_toolkits import mplot3d
```

### Fonts
Define fonts to use plot fonts that are similar to the classic LaTEX fonts used for the equations

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Load Colormap
Custom colormap from a png color image

```python
im = Image.open("../../assets/colormaps/watercolours.png")
im = np.array(im)
colormap = im[:,0,:] / 255.0
```

### Load Data
This data is a subset of a dataset hosted on [Kaggle](https://www.kaggle.com/datasets/mirichoi0218/insurance)

```python
data = pd.read_csv('../../data/healthcare_cost/insurance.csv')
columns = data.columns
x_orig = data.iloc[:,0].values
y_orig = data.iloc[:,-1].values / 1000.00
x = x_orig.copy() 
y = y_orig.copy()
```

### Create a color scheme for the categories in our data

```python
plot_colors = [
    colormap[50,:],
    colormap[200,:],
    colormap[0,:],
    colormap[100,:]]
plot_colors = np.array(plot_colors)
```

### Exploring the data
What ever data you might be using, it is always a good idea to plot a few summary statistics, distributions and/or individual input data points. There will always be surprises. There will always be a need for some decision making, in terms of normalizing, and because of outliers, missing values, etc. As we rely more and more on specialized toolboxes, we loose that feel for our data. Algorithms today might be black boxes, but at the very least we should know a thing or two about the data we provide it as input. The data we use here is already a good example of what you might learn, simply by plotting various views of the data. 

### Plotting the raw data
We should start by just plotting the raw data. The figure below plots total medical expenses (in dollars) as a function of age, which each dot representing a single data point. Already a few interesting effects can be observed. Across the board, it seems that medical expenses increase as a function of age, the effect we are hoping to predict. However, the data seems to be broken into 3 distincy bands, shifted relative to each other. How do we explain that?

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y,c='k')
plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=16,**csfont)
plt.yticks(range(0,60,10),labels=range(0,60,10),fontsize=16,**csfont)
plt.xlabel('age (years)',fontsize=18,**csfont);
plt.ylabel('medical expenses (dollars x1000)',fontsize=18,**csfont);
plt.grid()
plt.text(80,10,'(A)',fontsize=24,**csfont);
plt.text(80,30,'(B)',fontsize=24,**csfont);
plt.text(80,50,'(C)',fontsize=24,**csfont);
plt.axis((10,70,0,60))
plt.savefig('../../assets/figures/014/nb1.png',dpi=300,bbox_inches="tight")
```

### Plotting the same data, now color coded for tobacco use. 
Could the shift in expenses be the result of tobacco use? When we color code the data points based on the each individual self-identifying as a smoker, we indeed see that smokers appear shifted up, relative to their non-smokers. The relationship between age and medical expenses is about the same for smokers and non-smokers alike, with the difference being that smokers argue a higher baseline level of expenses, independent of age. However, this explains the differences between the upper and lower band of data pretty well, as smokers appear almost exclusively in the upper band, non-smokers almost exclusively in the lowers band. But what about the middle band?

```python
plt.figure(figsize=(6,6))
z = data['smoker'].values
classnames = ['non-smoker','smoker']
classes = np.unique(z)[::-1]
for i in range(len(classes)):
    ind = np.where(z == classes[i])[0]
    x_subset = x[ind]
    y_subset = y[ind]
    plt.scatter(x_subset,y_subset,color=plot_colors[i,:],label=classnames[i])
plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=16,**csfont)
plt.yticks(range(0,60,10),labels=range(0,60,10),fontsize=16,**csfont)
plt.xlabel('age (years)',fontsize=18,**csfont);
plt.ylabel('medical expenses (dollars)',fontsize=18,**csfont);
plt.grid()
L = plt.legend(bbox_to_anchor=(0.0, 1.20), loc='upper left',fontsize=16,frameon=False)
plt.setp(L.texts, family='Georgia')
plt.axis((10,70,0,60))
plt.savefig('../../assets/figures/014/nb2.png',dpi=300,bbox_inches="tight")    
```

### Plotting the same data, now color coded for sex
Can we explain the difference in medical expenses by the sex? Similar to the above, we plot the same data, but now color coded as a function of sex. Before doing so, we should acknowledge that both smoker and sex by themselves cannot explain the existence of the three bands, as both variables only have two levels. But whereas smoking at least seems to explain some of the band-like structure in our data (see above), gender does not: data points don't cluster into bands. 

```python
plt.figure(figsize=(6,6))
z = data['sex'].values
classes = np.unique(z)[::-1]
for i in range(len(classes)):
    ind = np.where(z == classes[i])[0]
    x_subset = x[ind]
    y_subset = y[ind]
    plt.scatter(x_subset,y_subset,color=plot_colors[i,:],s=50,label=classes[i])
plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=16,**csfont)
plt.yticks(range(0,60,10),labels=range(0,60,10),fontsize=16,**csfont)
plt.xlabel('age (years)',fontsize=18,**csfont);
plt.ylabel('medical expenses (dollars)',fontsize=18,**csfont);
plt.grid()
L = plt.legend(bbox_to_anchor=(0.0, 1.20), loc='upper left',fontsize=16,frameon=False)
plt.setp(L.texts, family='Georgia')
plt.axis((10,70,0,60))
plt.savefig('../../assets/figures/014/nb3.png',dpi=300,bbox_inches="tight")    
```

### Plotting the same data, now color coded for region
The final cagorical variables is region, which has four levels. Like sex, this seems to not explain any of the observed band-like structure. 

```python
plt.figure(figsize=(6,6))
z = data['region'].values
classes = np.unique(z)[::-1]
for i in range(len(classes)):
    ind = np.where(z == classes[i])[0]
    x_subset = x[ind]
    y_subset = y[ind]
    plt.scatter(x_subset,y_subset,color=plot_colors[i],s=50,label=classes[i])
plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=16,**csfont)
plt.yticks(range(0,60,10),labels=range(0,60,10),fontsize=16,**csfont)
plt.xlabel('age (years)',fontsize=18,**csfont);
plt.ylabel('medical expenses (dollars)',fontsize=18,**csfont);
plt.grid();
L = plt.legend(bbox_to_anchor=(0.0, 1.40), loc='upper left',fontsize=16,frameon=False);
plt.setp(L.texts, family='Georgia');
plt.axis((10,70,0,60))
plt.savefig('../../assets/figures/014/nb4.png',dpi=300,bbox_inches="tight")   
```

### Plotting the same data, now color coded number of children and BMI.
Perhaps we can explain the band like structure with the two other numerical variables: the number of children and the persons BMI. This seems unlikely, as the bandlike structure seems to indicate a categorical shift. If there was an effect of BMI, we would expect a more gradient like color coding. However, on close inspection, such a gradient is perhaps seen in the effect of BMI, although it is not very pronounced. The number of children, however, seems to have no effect of the amout of medical expenses argued. 

```python
plt.figure(figsize=(16,6))
numerical_columns = ['bmi','children']
for c,column in enumerate(numerical_columns):
    plt.subplot(1,2,1+c)
    plt.grid()
    z = data[column].values.copy()
    z[z > 35] = 35
    minval = np.min(z)
    maxval = np.max(z)
    z = z - minval
    z = z / (maxval - minval)
    z = (z * 255).astype(int)
    colors = colormap[z,:]
    plt.scatter(x,y,c=colors)
    if (c == 0): 
        plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=14,**csfont);
        plt.yticks(range(0,60,10),labels=range(0,60,10),fontsize=14,**csfont);        
        plt.xlabel('age',fontsize=18,**csfont);   
        plt.ylabel('medical expenses (dollars)',fontsize=18,**csfont);
    if (c == 1): 
        plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=14,**csfont);
        plt.yticks(range(0,60,10),labels=range(0,60,10),fontsize=14,**csfont);        
        plt.xlabel('age',fontsize=18,**csfont);   
    colorbar = plt.colorbar()  
    minval = np.floor(minval)
    maxval = np.ceil(maxval)
    crange = np.round(np.linspace(0,1,6),2)
    colorbar.set_ticks(crange)
    plt.axis((10,70,0,60))
    colorbar.set_ticklabels((minval + crange*maxval).astype(int),**csfont)
    colorbar.set_label(column,fontsize=16,**csfont);   
plt.savefig('../../assets/figures/014/nb5.png',dpi=300,bbox_inches="tight")    
```

When we zoom in on the lowest band, containing most of the data points, we notice that there is an effect of the number of children conceived on medical expenses. For BMI, however, 

```python
plt.figure(figsize=(16,6))
numerical_columns = ['bmi','children']
for c,column in enumerate(numerical_columns):
    plt.subplot(1,2,1+c)
    plt.grid()
    z = data[column].values.copy()
    z[z > 35] = 35
    minval = np.min(z)
    maxval = np.max(z)
    z = z - minval
    z = z / (maxval - minval)
    z = (z * 255).astype(int)
    colors = colormap[z,:]
    plt.scatter(x,y,c=colors)
    plt.xticks(range(10,80,10),labels=range(10,80,10),fontsize=14,**csfont);
    plt.yticks(range(0,18,2),labels=range(0,18,2),fontsize=14,**csfont);        
    plt.xlabel('age',fontsize=18,**csfont);   
    plt.ylabel('medical expenses (dollars)',fontsize=18,**csfont);
    colorbar = plt.colorbar()  
    minval = np.floor(minval)
    maxval = np.ceil(maxval)
    crange = np.round(np.linspace(0,1,6),2)
    colorbar.set_ticks(crange)
    colorbar.set_ticklabels((minval + crange*maxval).astype(int),**csfont)
    colorbar.set_label(column,fontsize=16,**csfont);   
    plt.axis((10,70,0,16))
plt.savefig('../../assets/figures/014/nb6.png',dpi=300,bbox_inches="tight")    
```

### Possible Interaction Effects?
The number of children, sex and the geographic region of the patient does not seem to drive health cost, or at least not visibly. Being a smoker and BMI seems to have an effect, but neither variables can explain the existence of the three bands observed in the data. This leaves the three-banned structure observed in the data unexplained by each variable when considered in isolation. 

It is possible that the effect is due to a variable not included in the data set. Variables represent different levels of insurance coverage, or some other demographic not listed. Alternatively, they could be an interaction effect between two variables. We can test the existence of interaction effects using linear regression, and we will return to this full data set later. For now, we will start with explaining a single output of interest (medical expenses) from a single input variable (age)
