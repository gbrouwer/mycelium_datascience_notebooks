### Import

```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Load Data

```python
data = pd.read_csv('../../data/healthcare_cost/insurance.csv')
columns = data.columns
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

### Clean Dataset for a univariate example

```python
ind0 = np.where(y > 17000)[0]
ind1 = np.where(X[:,0] < 46)[0]
ind2 = np.where(y > 10000)[0]
ind3 = np.intersect1d(ind1,ind2)
ind4 = np.union1d(ind3,ind0)
ind5 = range(X.shape[0])
ind5 = np.setdiff1d(ind5,ind4)
x = X[ind5,0]
y = y[ind5].astype('float')
```

### Random Sample for the Wiki

```python
np.random.seed(100)
idx = np.random.randint(0,x.shape[0],20)
x = x[idx]
y = y[idx] / 1000
meany = np.mean(y)
```

### Our Model

```python
def univariate_linear_model(x,a,b):

    y = x*a + b
    return y
```

### Compute Error

```python
def error(y,y_pred):
    
    return y - y_pred
```

### Plot Data with a single error point

```python
ypred = np.zeros((20,1)) + meany
x0 = 15
x1 = 65
y0 = meany
y1 = meany
plt.figure(figsize=(6,6));
plt.scatter(x,y,c='k');
plt.plot([x0,x1],[y0,y1],'k',zorder=100);
plotrange = list(range(x.shape[0]))
for index in [plotrange[0]]:
    plt.plot([x[index],x[index]],[y[index],ypred[index]],'r--')
    plt.scatter(x[index],ypred[index],c='red',zorder=100);
plt.grid();
plt.xlabel('age',fontsize=22,**csfont);
plt.ylabel('medical expenses (dollars x 1000)',fontsize=22,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.axis((10,70,0,15));
plt.savefig('../../assets/figures/014/rsquare_00.png', bbox_inches="tight")
```

### a = 0, b = mean(y)

```python
a = 0
b = np.mean(y)
w = [a,b]
stepsize = 10
ypred = x*a + b

#Left and right point of the linear regression line
x0 = 15
x1 = 65
y0 = x0*a + b
y1 = x1*a + b

SST = np.sum((y - np.mean(y))**2)
SSE = np.sum((y - ypred)**2)
SSR = np.sum((ypred - np.mean(y))**2)
rsquare = 1 - (SSE/SST)
print(a,b,rsquare)

plt.figure(figsize=(6,6));
plt.scatter(x,y,c='k');
plt.plot([x0,x1],[y0,y1],'k',zorder=100);
plotrange = list(range(x.shape[0]))
for index in plotrange:
    plt.plot([x[index],x[index]],[y[index],ypred[index]],'r--')
    plt.scatter(x[index],ypred[index],c='red',zorder=100);
plt.grid();
plt.xlabel('age',fontsize=22,**csfont);
plt.ylabel('medical expenses (dollars x 1000)',fontsize=22,**csfont);
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.axis((10,70,0,15));
plt.savefig('../../assets/figures/014/rsquare_01.png', bbox_inches="tight")
```

### Loss Landscape

```python
aval = np.linspace(-500,500,101)
bval = np.linspace(-4000,12000,161)
Xm,Ym = np.meshgrid(aval, bval)
XX = np.array([Xm.ravel(), Ym.ravel()]).T

lossfield = np.zeros((161,101))
for i in range(len(aval)):
    for j in range(len(bval)):
        ypred = aval[i]*x + bval[j]
        loss = np.sum((y - ypred)**2)
        lossfield[j,i] = loss / 1000
lossfield = lossfield / 1000 / 1000

plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(Xm,Ym,lossfield,cmap='YlGnBu', edgecolor='gray')
ax.view_init(elev=30., azim=120)
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
ax.set_zticks(np.linspace(0,20,5),labels=np.linspace(0,20,5),fontsize=14,**csfont);
ax.set_xlabel('\nage',fontsize=24,**csfont);
ax.set_ylabel('\n\nmedical expenses',fontsize=24,**csfont);
ax.set_zlabel('\nLoss',fontsize=24,**csfont);
plt.savefig('../../assets/figures/014/loss_surface_01.png', bbox_inches="tight")
plt.show();      

plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(Xm,Ym,lossfield,cmap='YlGnBu', edgecolor='gray')
ax.view_init(elev=60., azim=240)
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
ax.set_zticks(np.linspace(0,20,5),labels=np.linspace(0,20,5),fontsize=14,**csfont);
ax.set_xlabel('\nage',fontsize=24,**csfont);
ax.set_ylabel('\n\nmedical expenses',fontsize=24,**csfont);
ax.set_zlabel('\nLoss',fontsize=24,**csfont,rotation=-90);
plt.savefig('../../assets/figures/014/loss_surface_02.png', bbox_inches="tight")
plt.show();
```
