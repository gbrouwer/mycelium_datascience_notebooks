# Breast cancer dataset - Ordinary Least Squares
This demo uses the statistical approach outlined in 014.06, now using the Wisconscin Breast Cancer Dataset. Since this is a categorical dataset in nature (the outcome represents whether the tumor is malignant or benign), we instead use two input features as input and output to create a regression problem. Specifically, we will be using the feature radius_mean to predict both concave_points_mean and symmetry_mean

### Import

```python
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Fonts

```python
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Helvetica'}
```

### Load Data

```python
columns = []
with open('../../data/breast_cancer/wdbc.columns','r') as f:
    for line in f:
        columns.append(line.rstrip())
data = pd.read_csv('../../data/breast_cancer/wdbc.data',header=None)
data.columns = columns
```

### Split the data into input X and output y and drop the ID

```python
M = data[columns[2:]].values
columns = columns[2:]
y = data[columns[1]].values
```

### Recode y into a numerical variable

```python
classes = np.unique(y)
for i,item in enumerate(classes):
    indices = np.where(y == item)[0]
    y[indices] = i
```

### Select input and output data. 
We start with radius and compactness. In other words, can we predict the radius if we knew the compactness of a tumor?

```python
colx = 0
coly = 5
x = M[:,colx]
y = M[:,coly]
print(columns[colx])
print(columns[coly])
```

### Plot the data
Plotting the data shows a very strong correlation between variables. This should yield a reasonable fit. 

```python
plt.figure(figsize=(6,6));
plt.scatter(x,y,c='k');
plt.grid()
plt.axis((np.min(x),np.max(x),np.min(y),np.max(y)))
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel(columns[colx],fontsize=20,**csfont);
plt.ylabel(columns[coly],fontsize=20,**csfont);
plt.savefig('../../assets/figures/014/ols_approach_breastcancer_01.png', bbox_inches="tight");
```

### Step 1: Create Matrix

```python
X = np.ones((len(x),2))
X[:,0] = x
```

### Step 2: use OLD to compute the best fitting parameters a and b
Note that our input data matrix X has two columns. The first columns holds the data from 'radius_mean' variable, while the second is a column of 1s. This second column functions as the offset/intercept/bias encountered previously when we explored the statistical method for fitting linear regression models. For OLS, this offset is incoorperated into the input matrix X. Since its valuea are all set to 1, the resulting weights fitted to it can only describe an offetset term, not a slope. 

```python
W = np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)
a = W[0]
b = W[1]
```

### Plot the data overlayed with the fitted function

```python
xval = np.linspace(np.min(x),np.max(x),100)
yval = xval*a + b;
plt.figure(figsize=(6,6));
plt.plot(xval,yval,'r',linewidth=3);
plt.scatter(x,y,color=(0.0,0.0,0.0),s=35)
plt.grid()
plt.axis((np.min(x),np.max(x),np.min(y),np.max(y)))
plt.xticks(fontsize=14,**csfont);
plt.yticks(fontsize=14,**csfont);
plt.xlabel(columns[colx],fontsize=20,**csfont);
plt.ylabel(columns[coly],fontsize=20,**csfont);
plt.savefig('../../assets/figures/014/ols_approach_breastcancer_02.png', bbox_inches="tight");
```

### R-squared

To evaluate the quality of our fit, we compute the R-square. As we discussed previously, the r-squared is a value between 0 and 1, values close to 1 indicate a good fit of the model. Please note that R-square is a different metric than correlation or the slope of the fitted function. In the absence of a strong correlation between x and y, we can still have a good fit. The opposite is true as well: a strong correlation might still yield a poor fit if our model can't capture the true nature of the relationship between x and y. For example, the relationship might be highly nonlinear, so a linear model will not be able to capture that.  

To compute the R-square, we need to compute the SSR (sum of squares residual, or the amount of variance left over after we fit the function) and the SST (the total amount of variance in the data). 

```python
y_mean = np.mean(y)
SSR = np.sum((y - (a*x + b))**2)
SST = np.sum((y-y_mean)**2)
rsquared = 1 - (SSR/SST)
print(rsquared)
```
